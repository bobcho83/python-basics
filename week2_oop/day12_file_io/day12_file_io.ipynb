# 12일차: 파일 입출력 핵심 개념

1. 파일 열기(open)와 모드: 파일을 다룰 때는 "어떤 용도로 열 것인가"를 먼저 결정
- 'r'(Read): 읽기용 (파일이 없으면 에러)
- 'w'(Write): 쓰기용 (새로 만들거나 기존 내용을 다 지우고 새로 씀)
- 'a'(Append): 추가용 (기존 내용 뒤에 덧붙임)
2. 파일 읽기의 3가지 방법
- read(): 파일 전체 내용을 하나의 큰 글자 덩어리로 가져옴
- readline(): 딱 한 줄만 읽어옴
- readlines(): 모든 줄을 읽어서 리스트에 담아줌
3. with문: "뒷정리는 내가 할게!"(강력 추천): 원래는 파일을 열면 close()로 직접 닫아줘야 함. 하지만 with 문을 쓰면 파일을 다 쓰고 난 뒤 파이썬이 자동으로 안전하게 닫아줌. 밥님은 앞으로 이 방식을 주로 쓰게 됨.

```python
# [파일 입출력 - 업무 일지 시스템]

file_path = "daily_report.txt"

# 1.파일 쓰기 (with 문 사용)
with open(file_path, "w", encoding="utf-8") as f:
    f.write("--- 12일차 마케팅 리포트 ---\n")
    f.write("1. 파이썬 파일 입출력 공부 완료\n")
    f.write("2. 깃허브 12일 연속 잔디 심기 성공\n")
    print(f"'{file_path}' 파일에 기록을 완료했습니다.")

# 2.파일 추가하기 (Append 모드)
with open(file_path, "a", encoding="utf-8") as f:
    f.write("3. 내일은 예외 처리 공부 예정\n")

# 3.파일 읽기 (readlines 사용)
print("\n[저장된 파일 내용 읽기]")
with open(file_path, "r", encoding="utf-8") as f:
    lines = f.readlines()
    for line in lines:
        print(line.strip()) # strip()은 줄바꿈 문자를 제거해줌
```

**[핵심 요약]**
1. open(): 책꽂이에서 '책'을 꺼내서 펼치는 것과 같음
2. write(): 빈 종이에 연필로 '글씨'를 쓰는 것. 'w'는 새 종이고, 'a'는 이미 써진 종이 아래에 이어서 쓰는 것.
3. read(): 눈으로 책 내용을 '읽는' 것
4. with 문: 책을 다 읽고 나면 엄마가 "책 제자리에 꽂아둬~"라고 안 해도, 다 읽자마자 책이 자동으로 슈슉! 책꽂이로 돌아가는 마법 상자와 같음

#### 중급 핵심 문법: 구조화 된 데이터 다루기
1. JSON이란?: 데이터를 저장하거나 전송할 때 사용하는 가벼운 형식. 파이썬의 딕셔너리(Dictionary)와 거의 똑같이 생겨서 서로 변환하기가 아주 쉬움
2. json 모듈의 핵심 함수
- json.dump(): 파이썬 객체(리스트, 딕셔너리 등)을 파이렝 저장할 때 씀
- json.load(): 파일에 저장된 JSON 데이터를 파이썬 객체로 불러올 때 씀
- indent 옵션: 데이터를 저장할 때 줄 바꿈과 들여쓰기를 해줘서 사람이 보기 편하게 만들어줌

```python
import json

# 1.저장할 데이터 (파이썬 딕셔너리 형태)
campaign_config = {
    "campaign_name": "여름 신제품 프로모션",
    "budget": 5000000,
    "target_platform": ["Instagram", "Facebook", "Youtube"],
    "is_active": True
}

file_name = "config.json"

# 2.JSON 파일로 저장하기 (Dump)
with open(file_name, "w", encoding="utf-8") as f:
    # ensure_ascii=False : 한글이 깨지지 않게 저장
    # indent=4 : 보기 좋게 4칸 들여쓰기
    json.dump(campaign_config, f, ensure_ascii=False, indent=4)
    print(f"설정 파일 '{file_name}'이 생성되었습니다.")

# 3.JSON 파일 불러오기 (Load)
print("\n[설정 파일 불러오기 결과]")
with open(file_name, 'r', encoding="utf-8") as f:
    loaded_data = json.load(f)

    # 이제 불러온 데이터를 딕셔너리처럼 쓸 수 있음!
    print(f"캠페인명: {loaded_data['campaign_name']}")
    print(f"활성화 상태: {'진행 중' if loaded_data['is_active'] else '중단'}")
```

**[핵심 요약]**
1. JSON: 정보르르 담는 '정리함' 같은 것. 그냥 글자만 써 놓으면 나중에 찾기 힘들지만, "이름:Bob", "나이:43"처럼 이름표를 붙여서 정리함에 넣어두면 나중에 쏙쏙 꺼내 쓰기 좋음~
2. json.dump: 정리함에 물건을 차곡차곡 '넣는'행동
3. json.load: 정리함에서 물건을 다시 '꺼내는'행동
4. indent: 정리함 안의 물건들이 엉키지 않게 '칸막이'를 쳐주는 것. 칸막이가 있으면 우리가 보기에 훨씬 깔끔

#### 고급 핵심 문법: 데이터 분석의 기초
1. CSV 모듈 활용: CSV는 쉼표(,)로 구분된 표 형식의 데이터임. 파이썬의 csv모듈을 쓰면 엑셀 시트 한 줄을 리스트나 딕셔너리로 아주 쉽게 변환 할 수 있음
2. 예외 처리(try-except)연동: 파일 작업은 위험 요소가 많음 (파일이 없거나, 권한이 없거나 등)
- FileNotFoundError: 읽으려는 파일이 존재하지 않을 때 발생
- PermissionError: 파일이 이미 열려 있거나 읽기 권한이 없을 때 발생

```python
from os import write
import csv
import os

# 1.저장할 데이터 (표 형식: 리스트 안의 리스트)
performance_data = [
    ["날짜", "채널", "클릭수", "전환율"],
    ["2026-05-01", "Google", 1200, "3.5%"],
    ["2026-05-02", "Instagram", 850, "4.2%"],
    ["2026-05-03", "Facebook", 600, "2.1%"]
]

file_name = "marketing_data.csv"

# 2.csv 파일 쓰기
try:
    with open(file_name, "w", newline="", encoding="utf-8-sig") as f:
        writer = csv.writer(f)
        writer.writerows(performance_data)
    print(f"✅ '{file_name}' 저장 완료!")
    
    # 3.csv 파일 읽기 및 예외 처리
    print("\n[데이터 읽기 결과]")
    with open(file_name, "r", encoding="utf-8-sig") as f:
        reader = csv.reader(f)
        for row in reader:
            print(f"| {'|'.join(row)} |")

except FileNotFoundError:
    print("❌ 오류: 해당 파일을 찾을 수 없습니다. 경로를 확인하세요.")
except Exception as e:
    print(f"❌ 알 수 없는 오류 발생: {e}")
```

**[핵심 요약]**
1. csv: 데이터를 '바둑판 모양'의 표로 정리한 것. 엑셀과 친구라고 생각하면 됨
2. newline="" : 데이터를 쓸 때 줄 사이에 '빈 줄'이 생기지 않게 꽉꽉 채워주는 옵션
3. utf-8-sig: 엑셀에서 한글이 깨지지 않게 '한글 이름표'를 붙여주는 것
4. try-except: 이건 '안전모'임. 공사장에서 사고가 날 수 있듯이, 코딩할 때도 파일이 없는 사고가 날 수 있음. 그 때 프로그램이 죽지 않게 보호해주는 것.
