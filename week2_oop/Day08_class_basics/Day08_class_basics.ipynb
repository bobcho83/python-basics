# 클래스와 객체
1. 클래스(Class): 붕어빵을 찍어내는 '붕어빵 틀'임
2. 객체(Object): 그 틀에서 구워져 나온 실제 '붕어빵'임
3. 속성(Attribute): 붕어빵 안에 든 '팥'이나 '슈크림'같은 특징
4. 메서드(Method): 붕어빵이 하는 '따끈따끈하게 김 내뿜기'같은 행동

### 초급

```python
# [핵심 문법 정리]

# 1.클래스 정의(class 키워드)
class FishBread:
    pass # 일단 비워둘 때 사용

# 2.생성자(__init__)와 속성
def __init__(self, taste):
    self.taste = taste # 속성 설정

# 3.메서드(행동)
def show_info(self):
    print(f"{self.taste} 붕어빵입니다!")
```

```python
# [클래스와 객체 - 나만의 스마트폰 설계도]

# 1.클래스 정의 (설계도)
class SmartPhone:
    #2.속성 설정 (생성자)
    def __init__(self, brand, model):
        self.brand = brand # 브랜드
        self.model = model # 모델명
        self.is_on = False # 전원 상태 (기본값)

    # 3.메서드(행동)
    def power_on(self):
        self.is_on = True
        print(f"[{self.model}] 전원이 켜졌습니다. 📱")

    def show_info(self):
        status = "켜짐" if self.is_on else "꺼짐"
        print(f"기기 정보: {self.brand} {self.model} (상태: {status})")

# 4.객체 만들기(실제 물건 생성)
my_phone = SmartPhone("애플", "iphone 15 pro")
bob_phone = SmartPhone("삼성", "Galaxy S26")

# 5.행동 시키지
my_phone.power_on()
my_phone.show_info()
bob_phone.show_info()
```

**[핵심 요약]**

- class는 똑같은 물건을 여러개 만들 때 쓰는 틀임
- self는 "붕어빵 틀에서 나온 이 붕어빵"의 정보를 정확히 찾아가기 위해 꼭 필요
- 객체는 클래스라는 이름 뒤에()를 붙여서 만듬

### 중급

**중급 핵심 문법: 클래스의 진화**
1. 상송(Inheritance): 이미 만들어진 클래스의 기능을 그대로 물려받아 새로운 클래스를 만드는 기술
- 예) '가전제품' 클래스를 상속받아 '냉장고', '세탁기'클래스를 만들면 공통 기능(전원 켜기 등)을 다시 만들 필요가 없음
2. 메서드 오버라디이(Method Overriding): 부모에게 물려받은 기능을 내 입맛에 맞게 '재정의'하는 것
- 예) 부로가 준 "소리 내기" 기능을 자식 클래스인 '강아지'는 "멍멍",'고양이'는 "야옹"으로 바꾸는 식
3. 비공개 속성(Encapsulation 맛보기): 변수 이름 앞에 언더바 두 개(__)를 붙여서 클래스 외부에서 마음대로 값을 바꾸지 못하게 잠그는 기술임. 중요한 데이터(예: 잔액, 비밀번호)를 보호할 때 필수

```python
# [중급 - 클래스 상속과 데이터 보호]

# 1.부모 클래스 (기본 캠페인)
class Campaign:
    def __init__(self, name, budget):
        self.name = name
        self.__budget = budget # _ _를 붙여서 '비공개'로 설정 (보호!)

    def get_budget(self):
        return self.__budget
    
    def show_report(self):
        print(f"[{self.name}] 캠페인이 진행 중입니다. 예산: {self.__budget}원")

# 2.자식 클래스 (상속: SNS 캠페인)
class SNSCanmpaigh(Campaign):
    def __init__(self, name, budget, platform):
        # 부모 클래스의 생성자를 호출 (super)
        super().__init__(name, budget)
        self.platform = platform

    # 3.메서드 오버라이딩 (부모의 기능을 재정의)
    def show_report(self):
        print(f"[{self.platform} 전용] {self.name} 리포트입니다.")
        print(f"남은 예산은 {self.get_budget()}원입니다.")

# 4.객체 생성 및 실행
general_ad = Campaign("기본 광고", 1000000)
insta_ad = SNSCanmpaigh("신제품 홍보", 500000, "Instagram")

print("--- 알번 광고 리포트 ---")
general_ad.show_report()

print("\n--- SNS 광고 리포트 ---")
insta_ad.show_report()

# [주의] 아래코드는 에러 발생! (비공개 속성 보고)
# print(insta_ad.__budget)
```

**[핵심 요약]**
1. 상속: 아빠의 '요리 실력'을 물려받는 것과 같음. 처음부터 배울 필요 없이 아빠 실력을 그대로 쓰면 됨
2. 오버라이딩: 아빠는 '라면'을 끓여주지만, 나는 거기에 '해물'을 넣어서 더 맛있게 바꾸는 것과 같음. 이름은 같은 '라면'이지만 내용은 내 맘대로 바꿈
3. 비공개 속성: (_ _): 내 '비밀 일기장'에 자물쇠를 채우는 것. 밖에서는 함부로 볼 수 없고, 오직 나(class)만 열어볼 수 있게 만드는 보호 장치

### 고급

**고급 핵심 문법: 설계자의 도구**
1. 클래스 메서드(@classmethod) vs 정적 메서드(@staticmethod)
- 클래스 메서드: 개별 붕어빵이 아니라 '붕어빵 틀 자체'의 정보를 다룰 때 사용 (예: 지금까지 만든 붕어빵 개수 세기)
- 정적 메서드: 클래스 안에 있지만, 클래스 정보나 객체 정보가 딱ㅎ리 필요 없는 '순수 도구' 같은 함수
2. 추상 클래스(Abstract Class)
- "이걸 상속받는 자식들은 반드시 이 함수를 만들어야 해!" 라고 강제하는 설계도
- 실제 물건을 만들 순 없고, 오직 상속용 가이드라인 역할만 함
3. 매직 메서드(Magic Method / Dunder Method)
- __str__, __len__ 처럼 이름 앞뒤에 언더바 두 개가 붙은 특수 함수들
- 객체를 print()했을 때 어떻게 보일지, len()을 썼을 때 뭘 반환할 지 정의함

```python
from abc import ABC, abstractmethod # 추상 클래스를 위한 라이브러리

# 1.추상 클래스 (설계 지침서)
class Shop(ABC):
    @abstractmethod
    def open_store(self):
        # 상속받는 모든 지점은 반드시 이 함수를 직접 구현행햐 함!
        pass

# 2.실제 지점 클래스
class StarBread(Shop):
    total_sales = 0 # 모든 지점의 총 매출 (클래스 변수)

    def __init__(self, location):
        self.location = location

    # 추상 메서드 구현(강제 사항 완성!)
    def open_store(self):
        print(f"{self.location}점이 문을 열었습니다! 🍞")


    #클래스 메서드: 전체 매출 관리
    @classmethod
    def add_sales(cls, amount):
        cls.total_sales += amount
        print(f"전체 누적 매출: {cls.total_sales}원")

    # 매직 메서드: 객체를 출력할 때의 모습 정의
    def __str__(self):
        return f"이곳은 스타브레드 {self.location}점 입니다."

# 3.실행해보기
gangnam = StarBread("강남")
hongdae = StarBread("홍대")

gangnam.open_store()
print(gangnam) # __str__ 덕분에 예쁘게 출력

# 클래스 메서드를 통해 전체 관리
StarBread.add_sales(50000)
StarBread.add_sales(30000)
```

```python
# 실습 1. 학생 클래스

class Student:
    def info(self):
        print(f"이름: {self.name}, 학번: {self.student_id}, 점수: {self.score}")
    
student1 = Student()

student1.name = "철수"
student1.student_id = "2024001"
student1.score = 85

student1.info()
```

```python
# 다른 방법
class Student:
    def __init__(self, name, student_id, score):
        self.name = name
        self.student_id = student_id
        self.score = score

    def info(self):
        print(f"이름: {self.name}, 학번: {self.student_id}, 점수: {self.score}")

student1 = Student("철수", "2024001", 85)

student1.info()
```

```python
# 실습 2. 은행 계좌 클래스

class BankAccount:
    def __init__(self, account_number, blance):
        self.account_number = account_number
        self.blance = blance

    def deposit(self, amount):
        self.blance += amount
        print(f"잔액: {self.blance}")

    def whitdraw(self, amount):
        if self.blance >= amount:
            self.blance -= amount
            print(f"{amount}원이 인출되었습니다. 잔액은 {self.blance}원 입니다.")
        else:
            print("잔액이 부족합니다!")

account = BankAccount("123-456-789", 10000)

account.deposit(5000)
account.whitdraw(3000)
```
