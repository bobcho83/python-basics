# 예외 처리 핵심 개념
프로그램이 예상치 못한 상황에서 갑자기 멈추는 사고를 방지하는 '예외 처리(Exception Handling)'를 복습해봅시다. 튼튼한 집을 지을 때 소화기와 비상구를 설치하는 것과 아주 중요한 과정이다. 

#### 13일차 - 예외 처리 핵심 개념
1. try-except: "일단 해보고, 안 되면 일로 와!" 오류가 날 것 같은 코드를 try안에 넣고, 실제로 오류가 났을 때 처리할 내용을 except에 적음
2. 오류의 종료: 대표적인 사고들
- ValueError: 타입은 맞지만 값이 적절하지 않을 때 (예: 숫자를 넣어야 하는데 글자를 넣음)
- ZeroDivisionError: 숫자를 0으로 나누려고 할 때
- IndexError: 리스트에 없는 번호(인덱스)를 부를 때
3. finally: "무슨 일이 있어도 이건 꼭 해!" 오류가 나든 안 나든, 마지막에 무조건 실행되는 코드 (예: 파일 닫기)
4. raise: "이건 내가 정한 오류야!" 파이썬이 정한 오류 말고도, 개발자가 직접 오류를 일으켜서 경고를 줄 때 사용

```python
# [예외 처리 - 광고 성과 계산기]
def calculate_roi(profit, cost):
    # raise: 광고비가 음수면 일부러 오류 발생시키기
    if cost < 0:
        raise ValueError("광고비는 음수가 될 수 없습니다!")
    return (profit / cost) * 100

try:
    print("--- 광고 수익률(ROI) 계산기 ---")

    # 사용자가 숫자가 아닌 글자를 입력할 수도 있어요!
    p = int(input("이익을 입력하세요: "))
    c = int(input("광고비를 입력하세요: "))

    roi = calculate_roi(p, c)

# 여러 except로 상황별 대처하기
except ValueError as e:
    print(f"❌ 값 오류: {e} (숫자를 정확히 입력해 주세요)")

except ZeroDivisionError:
    print("❌ 나누기 오류: 광고비가 0원일 때는 수익률을 계산할 수 없습니다.")

except Exception as e:
    print(f"❌ 예상치 못한 사고 발생: {e}")

# 오류가 없을 때만 실행
else:
    print(f"✅ 계산 성공! 수익률은 {roi}% 입니다.")

# 무조건 실행 (청소하기)
finally:
    print("--- 계산기 프로그램을 종료합니다 ---")
```

**[핵심 요약]**
1. try: '도전!' 일단 한 번 해보는 것.
2. except: '안전 매트'임. 도전하다가 넘어졌을 때 아프지 않게 받아주는 역할을 함.
3. finally: 도전이 성공하든 실패하든 마지막에 '손 씻기'를 하는 것과 같음. 마무리는 꼭 해야 함.
4. raise: 축구 심판이 '휘슬'을 부는 것과 같음. "이건 규칙 위반이야!" 라고 직접 알리는 것.

#### [중급 핵심 문법: 사용자 정의 예외]
1. 나만의 예외 클래스 만들기: 파이썬의 모든 오류는 Exception 이라는 클래스를 부모로 두고 있음. 따라서 우리가 새로운 오류를 만들 때도 Exception 을 상속받기만 하면 됨.
2. 예외 발생시키기와 데이터 담기: 오류가 발생했을 때 단순히 메시지만 띄우는 게 아니라, 어떤 데이터 때문에 오류가 났는지 정보를 함께 담아서 보낼 수 있음.

```python
# [중급 - 사용자 정의 예외와 데이터 전달]

# 1.나만의 예외 클래스 정의 (Exception 상속)
class ForbiddenWordError(Exception):
    def __init__(self, word, message="사용 금지 단어가 포함되어 있습니다. "):
        self.word = word
        self.message = message
        # 부모 클래스(Exception)의 생성자 호출
        super().__init__(self.message)

# 2.광고 문구 검사 함수
def check_ad_text(text):
    forbidden_list = ["무조건", "100% 당첨", "도박"]

    for word in forbidden_list:
        if word in text:
            # 3.직접 정의한 예외 발생시키기
            raise ForbiddenWordError(word)
    print("✅ 광고 문구 검토 통과!")

# 4.실행 및 예외 처리
ad_draft = "이번 이벤트는 100% 당첨 보장입니다!"

try:
    check_ad_text(ad_draft)
except ForbiddenWordError as e:
    # 내가 담은 데이터(word)를 꺼내서 더 구체적으로 안내 할 수 있음
    print(f"❌ 광고 승인 거절: {e.message}")
    print(f"👉 문제가 된 단어: '{e.word}")
finally:
    print("--- 광고 검토 프로세스 종료 ---")
```

**[핵심 요약]**
1. 사용자 정의 예외: 학교에서 "복도에서 뛰기 금지" 같은 '우리 학교만의 규칙'을 만드는 것과 같음. 파이썬이 모르는 우리 팀만의 규칙을 새로 정해주는 것
2. super().__init__: 새로운 규칙을 만들더라도, 이게'오류'라는 사실은 변함 없어야 함. 그래서 "이건 파이썬이 알던 오류랑 같은 급이야!" 라고 Exception에 알려주는 과정
3. 데이터 담기: 그냥 "안 돼!"라고만 하는게 아나라, "네 가방에 든 '사탕' 때문에 안 돼!"라고 정확한 이유(단어)를 콕 집어서 알려주는 친절한 경고판과 같음

#### [고급 핵심 문법: 디버깅과 예외 전달]
1. traceback 모듈: "범인은 이 안에 있어!" 프로그램이 에러로 멈췄을 때, 에러가 발생한 파일 위치, 줄 번호, 함수 호출 순서를 모두 문자열로 추출할 수 있음.
2. 예외 체이닝(Exception Chaining): from / 하나의 오류를 처리하다가 다른 오류를 발생시켜야 할 때, 원래 어떤 오류 때문에 이 사달이 났는지 뿌리를 남겨두는 기술 

```python
import traceback

# 1.예외 체이닝을 위한 커스텀 에러
import traceback
class MarketingDataError(Exception):
    """마케팅 데이터 처리 중 발생하는 상위 에러"""
    pass

def process_raw_data(data):
    try:
        # 데이터가 숫자 형태가 아닐 때 에러 유도
        return int(data) * 1.1
    except ValueError as original_error:
        # 2.raise ... from ... : 원인(Value Error)을 밝히며 새 에러 던지기
        raise MarketingDataError("데이터 변환 중 치명적 오류 발생") from original_error

# 실행 및 로그 기록
data_list = ["100", "200", "300"] # '천원'에서 에러 발생 예상

for item in data_list:
    try:
        result = process_raw_data(item)
        print(f"결과: {result}")
    except MarketingDataError as e:
        print(f"❌ 비즈니스 로직 에러: {e}")

        # 3.traceback: 실제 에러의 상세 경로 출력 (디버깅용)
        print("\n--- [개발자용 상세 로그] ---")
        detail_log = traceback.format_exc()
        print(detail_log)

        # (옵션) 이 내용을 error_log.txt 파일에 저장할 수 있음!
```

**[핵심 요약]**
1. Traceback: 명탐정 코난이 '발자국'을 따라 범인을 찾는 것과 같음. 어디서부터 문제가 시작되었는지 거꾸로 타고 올라가서 범인(오류)를 찾아냄
2. 예외 체이닝(from): 범인을 잡았는데, 그 뒤에 '배후 세력'이 있는 것과 같음. "나는 지금 '마케팅 에러'라는 이름으로 잡혔지만, 사실 내 뒤에는 '숫자 변환 에러'라는 원인이 있었어!"라고 솔직하게 고백하는 것.
