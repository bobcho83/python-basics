# 다형성과 캡슐화

#### 11일차 핵심 개념 정리
1. 다형성(Polymorphism): "이름은 같은데 행동은 다르다!"는 뜻. 부모 클래스의 메서드를 자식들이 각자 자기 스타일로 재정의(Overriding)해서 사용하는 것.
2. 캡슐화(Encapsulation) & 접근 제어: 중요한 데이터를 함부로 만지지 못하게 상자(클래스) 안에 넣고 잠그는 것. 파이썬에서는 언더바(_)를 사용해 비밀 등급을 정함.
- Public(공개): 누구나 접근 가능(name)
- Protected(보호): 자식 클래스까지만 사용 권장(_name)
- Private(비밀): 클래스 내부에서만 접근 가능(__name)
3. Getter와 Setter: 잠긴 상자 안의 데이터를 안전하게 확인(Getter)하거나 수정(Setter)하기 위해 만든 통로.

```python
# [다형성과 캡슐화 - 스마트 광고 시스템]

# 1.부모 클래스 (다형성의 기초)
class Ad:
    def __init__(self, title):
        self.title = title
        # 2.캡슐화: 실비용은 __를 붙여 Private으로 보호! (접근 제어)
        self.__cost = 0

    # 3.Getter & Setter(@property 이용)
    @property
    def cost(self):
        return self.__cost

    @cost.setter
    def cost(self, value):
        if value < 0:
            print("오류: 광고비는 0보다 작을 수 없습니다.")
        else:
            self.__age = value # 안전하게 데이터 수정

    def run(self):
        print("광고를 송출합니다.")

# 4.다형성 구현: 같은 run() 메서드지만 겨로가가 다름
class VideoAd(Ad):
    def run(self):
        print(f"🎥영상 광고 '{self.title}'를 유튜브에 재생합니다.")

class BannerAd(Ad):
    def run(self):
        print(f"배너 광고 '{self.title}'를 네이버에 노출합니다.")

# 5.실행 및 테스트
ads = [VideoAd("AI 강의"), BannerAd("파이썬 챌린지")]

for a in ads:
    a.run() # 이건이 다형성!!!!

# 6.캡슐화 확인
my_ad = Ad("기초 마케팅")
my_ad.cost = 5000
print(f"현재 광고비: {my_ad.cost}")
```

**[핵심 요약]**
1. 다형성: '요리해!'라는 명령은 똑같지만, 중식 셰프는 짜장면을 만들고 일식 쎼프는 초밥을 만드는 것과 같음. 명령(메서드 이름)은 같아도 결과(행동)은 각자의 직업(클래스)에 따라 달라짐
2. 캡슐화: 내 '스마트폰 비밀번호'와 같음. 아무나 내 폰을 열어볼 수 없게 잠가두고, 오직 나(메서드)를 통해서만 확인할 수 있게 만드는 보호 장치
3. Getter/Setter: 금고 안의 돈을 직접 꺼낼 순 없지만, 은행원(Setter)에게 돈을 맡기면 은행원이 가짜 돈인지 검사하고 금고에 넣어주는 과정과 같음

#### 중급 핵심 문법: 유연하고 안전한 설계
1. 덕 타이핑(Duck Typing): 다형성의 정점임. "만약 어떤 새가 오리처럼 걷고, 오리처럼 꽥꽥거린다면, 그 새를 오리라고 부를 수 있다"는 철학
- 특징: 상속 관계가 아니더라도 메서드 이름만 같으면 똑같이 취급해주는 유연함임.
2. 추상 클래스와 다형성의 결합: 부모가 틀만 잡아놓으면, 자식들이 각자 다르게 행동하게 강제하는 방식
3. 데코레이터를 이용한 고급 캡슐화: @property를 단순히 값을 읽는 용도가 아니라, 여러 속성을 조합해서 새로운 값을 계산해내는 '가상 속성'으로 활용.

```python
# [중급 - 덕 타이핑과 고급 프로퍼티]

# 1.다형성(덕 타이핑) 예시
class KakaoTalk:
    def send(self, msg):
        print(f"💬 카카오톡 발송: {msg}")

class Line:
    def send(self, msg):
        print(f"📗 라인 발송: {msg}")

# 상속받지 않아도 'send' 메서드만 있으면 누구든 오케이!
def send_notification(service, message):
    service.send(message)

# 2.고급 갭슐화 (계산된 속성)
class Campaign:
    def __init__(self, name, budget, spent):
        self.name = name
        self.__budget = budget
        self.__spent = spent

    @property
    def remaining_budget(self):
        # 변수처럼 보이지만 내부적으로는 계산해서 결과를 알려줌!
        return self.__budget - self.__spent
    
    @property
    def status(self):
        if self.remaining_budget > 0:
            return "진행 중"
        return "예산 소진"

# 실행 및 테스트
kakao = KakaoTalk()
line = Line()

print("--- 다형성(덕 타이핑) 테스트 ---")
send_notification(kakao, "오늘의 뉴스레터 도착!")
send_notification(line, "오늘의 뉴스레터 도착!")

print("\n--- 고급 캡슐화 테스트 ---")
cp = Campaign("여름 세일", 1000000, 450000)
print(f"캠페인: {cp.name} | 상태: {cp.status}")
print(f"남은 예산: {cp.remaining_budget}원")
```

**[핵심 요약]**
1. 덕 타이핑: 축구 팀을 뽑을 때, "축구 선수 자격증"이 없어도 공을 잘 차기만 하면 팀원으로 뽑아주는 것과 같음. 자격(상속)보다 능력(메서드 이름)을 보는 거죠!
2. 계산된 속성: 내 가방에 1,000원짜리 3개, 500원짜리 2개가 있을 때, 가방한테 "총 얼마 있어?"라고 물어보면 가방이 알아서 계산해서 "4,000원!"이라고 대답해 주는 것과 같음. 내가 매번 계산할 필요가 없음!
