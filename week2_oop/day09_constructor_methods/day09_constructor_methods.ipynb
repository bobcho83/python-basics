# 생성자와 메서드

## 클래스의 내부 동작 (심화)
1. 생성자 (__init__): 자동 시작 버튼
- 객체가 탄생하는 순간 딱 한 번 자동으로 실행되는 특수 함수예요. 주로 객체의 초기 상태(이름, 나이 등)를 설정할 때
2. self: 클래스라는 설계도 안에서 "지금 만들어진 바로 그 객체"를 가리키는 손가락과 같음. self가 붙어야 "내 사물함", "내 점수"가 됨
3. 인스턴스ㅡ 변수 vs 클래스 변수 (중요!)
- 인스턴스 변수: 각 객체마다 개별적으로 가진 값 (예: 학생마다 다른 이름)
- 클래스 변수: 해당 클래스로 만든 모든 객체가 공통으로 공유하는 값 (예: 학교 이름)

```python
# [생성자와 변수의 종류 - 캐릭터 생성기]

class GameCharacter:
    # 1.클래스 변수 (모든 캐릭터가 공유하는 정보)
    game_name = "파이썬 온라인"
    total_characters = 0 # 생성된 전체 캐릭터 수

    # 2.생성자 (인스턴스 변수 설정)
    def __init__(self, nickname, job):
        self.nickname = nickname # 인스턴스 변수 (개별 닉네임)
        self.job = job # 인스턴스 변수 (개별 직업)

        # 캐릭터가 만들어질 때마다 전체 카운트
        GameCharacter.total_characters += 1
        print(f"\n--- {self.nickname}님이 입장하셨습니다 ---")

    # 3.메서드와 self 매개변수
    def show_info(self):
        # 인스턴스 변수를 쓸 때는 self를 꼭 붙여야 한다
        print(f"게임: {GameCharacter.game_name}") # 클래스 변수 접근
        print(f"닉네임: {self.nickname}")
        print(f"직업: {self.job}")
        print(f"현재 접속자 수: {GameCharacter.total_characters}")

# 4.객체 생성 및 실행
user1 = GameCharacter("전사Bob", "기사")
user1.show_info()

user2 = GameCharacter("마법사Karl", "위저드")
user2.show_info()
```

**중급 핵심 문법: 메서드의 3총사**
1. 인스턴스 메서드(Instance Method)
- 특징: 가장 흔히 쓰는 메서드로, 첫 번째 인자로 self를 받음
- 용도: 개별 객체의 데이터를 읽거나 수정할 대 사용
2. 클래스 메서드(Class Method)
- 특징: 첫 번째 인자로 클래스 자체인 cls를 받으며, 위에 @classmethod 라는 장식을 달아줌
- 용도: 모든 객체가 공유하는 클래스 변수를 다룰 때 사용
3. 정적 메서드(Static Method)
- 특징: self나 cls를 받지 않으며, 위에 @staticmethod를 달아줌
- 용도: 클래스 안에 있긴 하지만, 객체나 클래스의 정보가 굳이 필요 없는 '단순 계산 도구'같은 함수

```python
# [중급 - 메서드의 종류와 역할]

class Pizza:
    total_pizzas = 0 # 클래스 변수 (전체 판매량)

    def __init__(self, name, toppings):
        self.name = name
        self.toppings = toppings
        Pizza.total_pizzas += 1

    # 1.인스턴스 메서드: '나(self)'의 정보를 알려줌
    def show_pizza(self):
        print(f"{self.name} 피자 (토핑: {', '.join(self.toppings)})")

    # 2.클래스 메서드: '틀(cls)'의 정보를 관리함
    @classmethod
    def get_total_sales(cls):
        print(f"오늘 팔린 전체 피자 수: {cls.total_pizzas}판")
    
    # 3. 정적 메서드: 객체와 상관없는 공통 도구
    @staticmethod
    def calculate_price(size_inch):
        # 인치당 1000원이라고 가정하는 단순 계산기
        return size_inch * 1000

# 클래스 사용하기
my_pizza = Pizza("페퍼로니", ["치즈", "페퍼로니"])
your_pizza = Pizza("불고기", ["불고기", "양파"])

my_pizza.show_pizza() # 인스턴스 메서드 호출
Pizza.get_total_sales() # 클래스 메서드 호출 (클래스 이름으로!)

price = Pizza.calculate_price(12)
print(f"12인치 피자 예상 가격: {price}원")
```

**고급 핵심 문법: 똑똑한 속성 관리**
1. @property (게터, Getter)
- 메서드지만 변수처럼 호출할 수 있게 해주는 마법의 장식
- 예: user.get_age() 대신 user.age로 깔끔하게 정보를 가져올 수 있음
2. @속성명.setter (세터, Setter)
- 변수에 값을 넣을 때 검사(유효성 검사)를 할 수 있게 해줌
- 예: 나이에 -5를 넣으려고 하면 "안 돼!"라고 막아주는 필터 역할
3. 정보 은닉(Information Hiding)
- 변수 앞에 _ _ 를 붙여 외부에서 직접 수정하지 못하게 숨기고, 오직 내가 만든 메서드(Setter)를 통해서만 값을 바꿀 수 있게 강제하는 설계 기법

```python
# [고급 - @property를 활용한 안전한 데이터 관리]

class Member:
    def __init__(self, name, age):
        self.name = name
        # _ _를 붙여 내부 변수로 숨김
        self.__age = age

    # 1.Getter: 변수처럼 값을 읽어오게 함
    @property
    def age(self):
        return self.__age

    # 2.Setter: 값을 넣을 때 잘못된 값이 들어오는지 검사함
    @age.setter
    def age(self, value):
        if value < 0:
            print("오류: 나이는 0보다 작을 수 없습니다!")
        else:
            self.__age = value
            print(f"나이가 {value}(으)로 변경되었습니다.")

# 사용하기
m = Member("Bob", 25)

# ()없이 변수처럼 읽기 (@property 덕분!)
print(f"회원 이름: {m.name}, 나이: {m.age}")

# 값 수정하기 (@age.setter 호출)
m.age = 30 # 정상 수정
m.age = -5 # 오류
```

**[핵심 요약]**
1. @property(Getter): 우리 집 창문임. 밖에서도 안이 어떻게 생겼는지 볼 수는 있지만, 직접 들어오지는 못하게 보호해줌
2. @setter: 우리 집 '대문 보안관' 아무나 들어오지 못하게 검사하고, 이상한 사람(잘못된 데이터)은 "못 들어가!" 하고 쫓아내는 역할
3. __age: 금고 속에 든 '비밀 통장' 대놓고 꺼내두면 누가 훔쳐갈 수 있으니까 꽁꽁 숨겨두고, 오직 보안관(Setter)을 통해서만 돈을 넣게 만든 것
