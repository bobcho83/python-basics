# 상속(Inheritance)
오늘 주제인 '상속(Inheritance)'은 객체지향 프로그래밍의 재사용성을 극대화하는 핵심 기술

### 상속: 클래스의 가계도
상속은 이미 잘 만들어진 클래스(부모)의 능력을 새로운 클래스(자식)가 그대로 이어받는 것
- 부모 클래스(Parent/Super Class): 공통적인 특징을 가진 기본 클래스 (예: '자동차')
- 자식 클래스(Child/Sub Class): 부모의 기능을 물려받으면서 자신만의 특징을 추가한 클래스 (예: '전기차, '스포츠카')

부모 클래스인 Car를 만들고, 이를 상속받는 ElecticCar를 설계. Super()의 사용법에 집중!

```python
# [상속 - 자동차와 전기차의 관계]

# 1.부모 클래스 정의
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def drive(self):
        print(f"{self.brand} {self.model}이(가) 달립니다. 부릉부릉!")

# 2.자식 클래스 정의(괄호 안에 부모 클래스 이름을 넣어요)
class ElectricCar(Car):
    def __init__(self, brand, model, battery_size):
        # 3.super(): 부모 클래스의 생성자를 호출해서 brand, model을 세팅!
        super().__init__(brand, model)
        self.battery_size = battery_size # 자식 클래스만 가진 새로운 속성

    # 4.부모의 메서드를 자식에게 맞게 재정의 (오버라이딩)
    def drive(self):
        print(f"{self.brand} {self.model}이(가) 조용히 달립니다. 위이잉~")
    
    def charge(self):
        print(f"{self.battery_size}kWh 배터리를 충전 중입니다.")

# 5.실행 및 확인
my_car = Car("현대", "아반떼")
my_ev = ElectricCar("테슬라", "Model 3", 75)

my_car.drive() # 부모의 기능 실행
my_ev.drive() # 자식이 재정의한 기능 실행
my_ev.charge() # 자식만 가진 기능 실행
```

### 중급 핵심 문법: 복잡한 상속의 세계
1. 다중 상속(Multiple Inheritance): 하나의 자식 클래스가 두 개 이상의 부모 클래스로부터 기능을 물려받는 것
- 예: '스마트폰'클래스는 '전화기'와 '카메라'클래스 모두를 상속받을 수 있음.
2. MRO(Method Resolution Order): 상속의 우선순위
- 여러 부모가 같은 이름의 메서드를 가지고 있다면 누구 것을 먼저 써야 할까? 파이썬은 MRO라는 규칙에 따라 순서를 결정. 클래스명.mro()를 통해 그 순서를 확인함


```python
# [중급 - 다중 상속과 우선순위]

# 1.첫 번째 부모: 메일 발송 기능
class EmailSender:
    def send(self, message):
        print(f"📧 이메일 발송: {message}")

# 2.두 번째 부모: SMS 발송 기능
class SMSSender:
    def send(self, message):
        print(f"📱 SMS 발송: {message}")

# 3.자식 클래스: 다중 상속(괄호 안에 두 클래스를 쉼표로 연결)
class MarketingBot(EmailSender, SMSSender):
    def announce(self):
        print("--- 광고 캠페인 시작 ---")
        # 부모가 둘 다 send()를 가지고 있다면?
        # 앞에 적어준 EmailSender의 send가 먼저 호출 됨!

# 4.실행 및 우선순위 확인
bot = MarketingBot()
bot.announce()

# MRO 확인: 파이썬이 메서드를 찾는 경로를 보여줌
print(f"\n 탐색 순서(MRO): {MarketingBot.mro()}")
```

### 고급 핵심 문법: 강제성을 부여하는 설계도
1. 추상 클래스(Abstract Class)
- 정의: 구현되지 않은 메서드를 포함하고 있어, 직접 객체를 만들 수 없는 클래스
- 목적: 상속받는 자식 클래스들에게 "이 메서드는 너희가 반드시 직접 만들어야 해!"라고 가이드라인을 주는 것.
2. 추상 메서드(@abstractmethod)
- 자식 클래스에서 반드시 오버라이딩(재정의)해야 하는 메서드. 이걸 만들지 않으면 자식 클래스도 객체를 만들 수 없음.

```python
from abc import ABC, abstractmethod

# 1.추상 클래스 정의 (ABC를 상송받아야 함)
class AdPlatform(ABC):
    def __init__(self, name):
        self.name = name

    # 2.추상 메서드 (자식이 반드시 구현해야 함)
    @abstractmethod
    def post_ad(self, content):
        pass

    # 일반 메서드 (공통 기능으로 물려줌)
    def welcome(self):
        print(f"[{self.name}] 광고 플랫폼에 오신 것을 환영합니다.")

# 3.자식 클래스: 구글 광고
class GoogleAd(AdPlatform):
    def post_ad(self, content):
        print(f"🌏 google 검색 결과에 광고 노출: {content}")

# 4.자식 클래스: 페이스북 광고
class FacebookAd(AdPlatform):
    def post_ad(self, content):
        print(f"🧍🏻🧍🏻Facebook 타임라인에 광고 노출: {content}")

# 실행
platform = [GoogleAd("구글"), FacebookAd("페이스북")]

for p in platform:
    p.welcome()
    p.post_ad("신학기 맞이 파이썬 강의 할인")
    print("-" * 30)

#[주의] 추상 클래스는 직접 만들 수 없음!
# test = AdPlatform("에러발생") # <- TypeError 발생
```
