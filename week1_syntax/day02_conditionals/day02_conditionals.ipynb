# 조건문 if / elif / else

### 초급

#### 1. 조건문(if) : 프로그램을 만들다 보면 '만약 ~라면 A를 하고, 아니면 B를 해!' 라고 정해줘야 할 때.
@ 신호등으로 이해하기
- if(만약) : '초록불이면?' -> "길을 건너요."
- elif(그게 아니고 만약) : "노란불이면?" -> "기다려요"
- else(그 외 나머지 전부) : "빨간불이면?" -> "멈춰요."

파이썬에서 조건문을 쓸 때 꼭 지켜야 할 2가지 약속
1) 콜론(:) : 조건이 끝나는 부분에 꼭 찍어줘야함. "이제 규칙 시작한다" 라는 신호.
2) 들여쓰기(스페이스 4번 or tab) : 조건에 해당할 때 실행할 코드는 안으로 쑥 밀어넣어야함. 파이썬은 이 간격을 보고 "아, 이 코드가 if에 연결된 거구나!" 라고 이해합니다.

```python
# [초급 조건문 - 놀이공원 입장료 판별기]
# 1) 사용자로부터 나이를 입력받아요.
# input()으로 받은 값은 '문자(str)'라서 '숫자(int)'로 바꿔줘야함.
age = int(input("나이가 어떻게 되세요? (나이를 입력해주세요!)"))

# 2) 조건에 따라 다른 메세지 출력
if age >= 20: #20세 이상일 때
    print("성인입니다. 티켓 가격은 50,000원입니다.")

elif age >= 14: #14세 이상일 때
    print("청소년입니다. 티켓 가격은 30,000원입니다.")

elif age >= 8: #8세 이상일 때
    print("어린이입니다. 티켓 가격은 10,000원입니다.")

else: # 그 외(7세 이하)
    print("미취학 아동입니다. 티켓은 무료입니다.")

# TIP : 파이썬에서 "크거나 같다"를 >= 로 씀.
```

[핵심 요약]
- if : 제일 처음 검사하는 조건
- elif : 위에서 '아니요'라고 했을 때, 다음으로 물어보는 질문
- else : 앞에 나온 모든 질문에 해당하지 않을 때 마지막 단계

### 중급
중급 과정의 핵심은 단순히 "A면 B다"를 넘어, 여러 개의 조건을 동시에 따지거나(논리 연산자), 조건문 안에 조건문을 넣은(중첩 조건문)등 더 복잡한 상황을 똑똑하게 처리하는 것.

1. 중급 핵심 문법 : 더 똑똑한 판별법
1) 논리 연산자 (and, or, not) : 조건을 하나만 따지는 게 아니라, 두 개 이상을 조합할 때 사용.
- and : "A도 맞고 B도 맞아야 해!" (두 조건 모두 참일 때)
- or : "A나 B 중 하나만 맞아도 돼!" (둘 중 하나만 참이어도 실행)
- not : "이 조건이 아니어야 해!" (참을 거짓으로, 거짓을 참으로 반전)

2) 중첩 조건문(Nested if) : 조건문 안에 또 다른 조건문을 넣는 방식. 큰 범위를 먼저 거르고, 그 안에서 세부적인 내용을 다시 나눌 때 사용.

3) 조건부 표현식(Ternary Operator) : 아주 간단한 if-else 문을 딱 한 줄로 줄여 쓰는 세련된 방법.
- 형식 : 참일 때 값 if 조건 else 거짓일 때 값

```python
# [중급 조건문 - 스마트 로그인 시스템]

# 1.초기 데이터
stored_id = "admin"
stored_pw = "1234"
is_premium = True

# 2.사용자 입력
user_id = input("아이디를 입력하세요: ")
user_pw = input("비밀번호를 입력하세요: ")

# 3.논리 연산자와 중첩 조건문 활용
if user_id == stored_id and user_pw == stored_pw:
    print("✅로그인이 완료되었습니다.")

    # 중첩 조건문 : 로그인 성공한 사용자 내에서 다시 등급 나누기
    if is_premium:
        print("💎 프리미엄 회원님, 환영합니다! 모든 기능을 사용하실 수 있습니다.")
    else:
        print("👶 일반 회원님, 환영합니다.")
    
elif user_id != stored_id:
    print("❌ 아이디가 일치하지 않습니다.")

else:
    print("🔑 비밀번호가 틀렸습니다.")

# 4.조건부 표현식(한 줄로 결과 만들기)
# 메시지가 길어질 때 아주 유용해요!
status = "접근 승인" if is_premium or user_id == "admin" else "접근 제한"
print(f"현재 사용자 상태: {status}")
```

### 고급
고급 단계의 조건문은 단순히 if를 쓰는 것을 넘어, "어떻게 하면 더 깔끔하고, 빠르고, 읽기 쉬운 코드를 만들 것인가?"에 집중합니다. 파이썬 전문가들이 즐겨 쓰는 세련된 기법 3가지를 소개하겠습니다.

1. Match-Case 문 (Python 3.10+) : 다른 언어의 switch-case와 비슷하지만 훨씬 강력하다. 단순히 값을 비교하는 게 아니라, 데이터의 패턴을 분석해서 처리한다. if-elif가 너무 길어질 때 사용하면 코드가 예술작품처럼 깔끔해짐.
2. 딕셔너리 매핑 (Dictionary Mapping) : 조건에 따라 특정 함수나 값을 실행해야 할 때, if 대신 dict를 사용하는 기법이다. 코드가 간결해지고 실행 속도도 빨라짐.
3. 올인원 조건 검사 (all(), any()) : 리스트 안에 있는 수많은 조건이 모두 참인지(all), 아니면 하나라도 참인지(any)를 한 줄로 검사함.

```python
# [고급 조건문 - 패턴 매칭과 효율적 매핑]

# 1.Match-Case 문 활용 (패턴 매칭)
# 다양한 형태의 명령어를 똑똑하게 처리합니다
def process_command(command):
    match command.split():
        case ["quit"]:
            return "프로그램을 종료합니다."
        case ["move", direction]:
            return f"{direction} 방향으로 이동합니다."
        case ["attack", target, "with", weapon]:
            return f"{target}를 {weapon}로 공격합니다."
        case _:
            return "알 수 없는 명령입니다."

print(process_command("move North"))
print(process_command("attack Zombie with Sword"))

# 2.딕셔너리 매핑 기법 (if-elif 대체)
# 복잡한 조건문 없이 결과값을 즉시 찾습니다.
def get_user_grade(score):
    grade = {9: "A", 8: "B", 7: "C"}
    return grade.get(score // 10, "F")

print(f"내 등급은: {get_user_grade(85)}")

# 3.all()과 any() 활용
# 여러 조건을 한 번에 검사할 때 최고!!
requirments = [
    True, # 이메일 인증 완료
    True, # 휴대폰 인증 완료
    False # 마케팅 수신 완료
]

if all(requirments):
    print("✅ 모든 필 수 조건을 만족했습니다!!")
else:
    print("⚠️ 아직 완료되지 않은 필수 조건이 있습니다.")
```

1) 고급 핵심 문법: 코드를 더. 견고하게 만들기
1. getattr()와 operator를 활용한 동적 조건 처리 : 조건에 따라 실행할 함수나 속성을 '이름(문자열)'으로 직접 찾아 실행하는 방식. if문을 수백 개 써야 할 상황을 단 한줄로 해결.
2. 서킷 브레이킹(Circuit Breaking) 연산 : 파이썬의 and와 or은 앞의 결과만 보고 뒤를 실행할지 말지 결정하는 '단축 평가(Short-circuit)'를 한다. 이를 이용해 에러를 방지하는 고급 기법을 쓸 수 있음.
3. 리스트/딕셔너리 컴프리헨션 내의 조건문 (Nested Conditions) : 데이터를 필터링하면서 동시에 변환까지 하는 기술.

```python
# [고급 조건문 확장 - 동적 처리와 단축 평가]

import operator

# 1.동적 연산 처리 (if-elif의 완전한 제거)
# 사용자가 입력한 기호(+, -)에 따라 계산을 수행.
ops = {
    "+": operator.add,
    "-": operator.sub,
    "*": operator.mul,
    "/": operator.truediv
}

def smart_calc(a, op, b):
    # 연산자가 사전에 있으면 실행, 없으면 "지원 안 함" 출력
    return ops.get(op, lambda x, y: "지원하지 않는 연산자")(a, b)

print(f"계산 결과: {smart_calc(10, '*', 5)}") # 결과: 50

# 2.단축 평가(Short-circuit)를 활용한 안전한 코드
# 데이터가 None이거나 비어있을 때 발생할 수 있는 에러를 방지
user_data = {"name": "Bob"} # 혹은 None

# user_data가 존재(True)할 때만 .get()을 실행함
user_name = user_data and user_data.get("name") or "익명 사용자"
print(f"로그인 유저: {user_name}")

# 3.컴프리헨션 안의 다중 조건 (If-Else)
# 데이터 리스트를 한 번에 가공
# 80점 이상은 'pass', 90점 이상은 'honors', 나머지 'fail'
scores = [98, 75, 82, 60, 91]
results = [
    "honors" if s >= 95 else "pass" if s >= 80 else "fail"
    for s in scores
]
print(f"성적 결과 리스트: {results}")
```

[요약]
1. 동적 연산 : "더하기는 1번방, 빼기는 2번방!" 이라고 방 번호표(사전)를 미리 붙여두고, 손님이 번호표만 내밀면 바로 그 방으로 안내
2. 단축 평가 : "냉장고에 우유가 있어?(and) -> 있으면 가져와!" 순서로 생각. 우유가 없으면 아예 꺼내러 가지도 않아서 실수로 빈 손을 휘저을 일 없음
3. 다중 조건 컨프리헨션 : 공장에서 물건을 분류할 때, 컨베이어 벨트 한 줄 위에서 '빨간 건 왼쪽, 파란 건 오른쪽, 나머지는 아래로!' 라고 아주 빠르게 분류하는 것과 같음
