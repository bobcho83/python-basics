### 중급

```python
# 중급 자료형 활용 - 데이터 필터링과 가공

# 1.샘플 데이터 (성과 데이터 가정)
campaign_results = [
    {"name": "SNS 광고", "Clicks": 150, "Cost": 500},
    {"name": "이메일 뉴스레터", "Clicks": 80, "Cost": 100},
    {"name": "검색 광고", "Clicks": 200, "Cost": 800},
    {"name": "인플루언서 협찬", "Clicks": 300, "Cost": 1500}
]

# 2.리스트 컴프리헨션(List Comprehension)
# 클릭 수가 100번 이상인 캠페인의 이름만 뽑아서 리스트 만들기
successful_campaigns = [c["name"] for c in campaign_results if c["Clicks"] >= 100]

# 3.딕셔너리 안전하게 접근 (.get)
# 데이터에 'conversion' 항목이 없을 수 있으니 기본값을 0으로 설정
for camp in campaign_results:
    conv = camp.get("conversion", 0)
    # f-string 슬라이싱 활용 : 캠페인 이름의 앞 두 글자만 출력
    print(f"[{camp['name'][:2]}] 클릭: {camp['Clicks']}, 전환: {conv}")

# 4.데이터 가공 챌린지
# 캠페인별 '클릭당 비용(cpc)'을 계산해서 새로운 딕셔너리에 저장
cpc_report = {c["name"]: c["Cost"] / c["Clicks"] for c in campaign_results}

print("\n--- 캠페인 CPC 보고서---")
for name, cpc in cpc_report.items():
    print(f"{name}: {cpc:.2f}원") # :.2f 소수점 둘째자리까지 출력
```

# 고급

1. 고급 핵심 문법: 데이터의 깊은 이해
- 가번(Mutable) vs 불변(Immutable) 객체의 비밀 : 파이썬은 변수가 값을 가지는게 아니라 주소를 가리킴. int, str은 한번 만들면 못 바꾸는 불변, list, dict는 안의 내용을 바꿀 수 있는 가변. 이 차이를 모르면 나중에 데이터가 꼬이는 큰 버그를 만날 수 있음
- 팩킹과 언팩킹 (Packing & Unpacking) : 리스트나 딕셔너리에 담긴 데이터를 한 번에 여러 변수로 흩뿌리거나, 반대로 여러 변수를 하나로 묶는 기술. 특히 *args와 **kwargs를 이해하는 출발점이 됨.
- 제너레이터(Generator) 표현식 : 리스트 컴프리헨션은 모든 데이터를 메모리에 한꺼번에 올리지만, 제너레이터는 사용할 때만 하나씩 만듬. 데이터가 100만개라면 리스트보다 훨씬 빠르고 메모리도 적게 씀.

```python
# 고급 자료형 활용 - 언팩킹과 효율적 데이터 처리

# 1.딕셔너리 병합과 언팩킹 (Python 3.9+ 기법)
default_setting = {"theme": "dark", "font": "Nanum", "debug": False}
user_setting = {"font": "Fira Code", "debug": True}

# ** 두 개를 사용해 두 딕셔너리를 합침. 뒤에 오는 것이 우선순위를 가짐
merged_setting = {**default_setting, **user_setting}
print(f"설정 적용 결과: {merged_setting}")

# 2.리스트 언팩킹의 정수
scores = [95, 80, 85, 70, 100]
first, *middle, last = scores # 첫 번째, 마지막, 그리고 나머지를 분리!
print(f"첫 점수: {first}, 마지막 점수: {last}, 중간 점수들: {middle}")

# 3.제너레이터(Generator)로 메모리 아끼기!
# 리스트 [] 대신 소괄호()를 쓰면 제너레이터가 됨.
# 1억개의 숫자가 있어도 메모리를 거의 차지하지 않아요.
huge_data = (x**2 for x in range(1000000))
print(f"제너레이터 객체 확인: {huge_data}")
print(f"첫 번째 결과값: {next(huge_data)}") # 필요할 때만 하나씩 꺼내기

# 4.고급 챌린지: 딕셔너리 컴프리헨션
# 리스트를 바탕으로 5글자 이상인 단어만 '단어: 길이' 형태의 딕셔너리로 만듬.
words = ["python", "ai", "marketing", "data", "developer"]
word_map = {w: len(w) for w in words if len(w) >= 5}
print(f"필터링된 단어 사전: {word_map}")
```

#### 요약
- 언팩킹(*) : 꽉 묶인 선물 꾸러미를 한꺼번에 풀어서 바닥에 촤르르 펼쳐놓는 것과 같음.
- 가변/불변 : 숫자나 글자는 조각상처럼 한 번 만들면 못 고치지만, 리스트나 사전은 레고 블록처럼 언제든 부수고 다시 조립 가능.
- 제너레이터 : 과자를 한 봉지 다 뜯어놓는 게(리스트) 아니라, 배고플 때만 하나씩 나오는 간식 기계를 만듬. 공간을 아주 작게 차지!!!
