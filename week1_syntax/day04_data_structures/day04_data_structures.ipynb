# 리스트(List)와 딕셔너리(Dictionary)

데이터를 묶어서 관리하는 파이썬의 가장 강력한 도구인 리스트와 딕셔너리를 정복한다.

1. 리스트(List)와 딕셔너리(Dictionary)
- 리스트 : 여러 개의 물건을 칸마다 담아둔 '기차'다. 몇 번째 칸인지가 중요.
- 딕셔너리 : 이름표가 붙은 '사물함'이다. 번호 대신 이름표(키)를 보고 물건(값)을 찾음.

### 초급

[문법 정리]
1) 리스트(List) - 순서대로 담기 / 리스트는 순서가 0번부터 시작한다는 점을 꼭 기억!
- append(값) : 기차 맨 뒤에 칸을 추가
- pop() : 맨 뒤의 칸을 떼어냄
- insert(번호, 값) : 원하는 위치에 칸을 쏙 키워 넣음
- remove(값) : 이름이 같은 칸을 찾아 지움
- 슬라이싱 [시작:끝] : 기차의 특정 구간만 툭 잘라냄 (끝 번호는 포함 안 됨!)

2) 딕셔너리(Dictionary) - 이름표로 찾기
- keys() : 모든 이름표(키)만 모아서 보여줌
- values() : 모든 알맹이(값)만 모아서 보여줌
- items() : 이름표와 알맹이 쌍을 세트로 보여줌

```python
# [초급 - 리스트와 딕셔너리 마스터]

# 1.리스트 연습: 간식 기차 만들기
snacks = ["초코파이", "사탕", "젤리"]
snacks.append("과자") # 뒤에 과자 추가
snacks.insert(1, "초콜릿") # 1번 자리에 끼어넣기
snacks.pop() # 마지막 '과자' 삭제

print("--- 우리 반 간식 기차 ---")
print(f"전체 간식 {snacks}")
print(f"앞에서 2개만 쏙! (슬라이싱): {snacks[0:2]} ")

# 2.딕셔너리 연습: 친구들의 최애 간식 사물함
favorite_snacks = {
    "철수": "사탕",
    "영희": "초콜릿",
    "Bob": "엿"
}

print("\n--- 최애 간식 사물함 ---")
# 새로운 친구 추가
favorite_snacks["민수"] = "쿠키"

# 딕셔너리 메서드 활용
print(f"누구의 이름표가 있나요? {list(favorite_snacks.keys())}")
print(f"어떤 간식들이 있나요? {list(favorite_snacks.values())}")

# 반복문으로 이름표와 간식 같이 부르기
for name, snack in favorite_snacks.items():
    print(f"{name} 친구는 {snack}을/를 좋아해요!")
```

[요약]
 - 리스트는 순서가 중요할 때 사용 (대기줄, 성적 순위)
 - 슬라이싱[0:3]은 0, 1, 2 번까지만 가져온다는 뜻 (0은 시작 위치, 3은 수량)
 - 딕셔너리는 데이터의 의미가 중요할 때 사용 (전화번호부, 사전)

### 중급

중급 단계에서는 리스트와 딕셔너리를 따로 쓰는게 아니라, 하나로 합쳐서 복잡한 데이터를 구조적으로 다루는 법 확인

1. 중급 핵심 문법 : 데이터 구조화
- 리스트 안의 딕셔너리 (복합 구조) : 우리가 흔히 보는 주소록, 상품 목록, 게임 캐릭터 정보 등은 모두 이 구조
- 리스트 컴프리헨션 (List Comprehension) : 반복문과 조건문을 한 줄로 합쳐서 리스트를 새로 만드는 기술
- 딕셔너리 .get()과 update() // .get(key, 값) : 키가 없어도 에러 대신 우리가 정한 기본값을 돌려줌 // update() : 여러 데이터를 한꺼번에 수정하거나 추가할 때 사용

```python
# [중급 - 리스트와 딕셔너리 복합 활용]

# 1.복합 데이터 구조(리스트 안에 딕셔너리)
students = [
    {"name": "철수", "score": 85, "grade": "B"},
    {"name": "영희", "score": 95, "grade": "A"},
    {"name": "민수", "score": 70, "grade": "C"},
    {"name": "Bob", "score": 100, "grade": "A+"}
]

# 2.리스트 컴프리헨션: 성적이 90점 이상인 '이름'만 뽑아내기
honor_students = [s["name"] for s in students if s["score"] >= 90]
print(f"우수 학생 명단: {honor_students}")

# 3.딕셔너리 데이터 수정 및 안전하게 가져오기
new_info = {"score": 98, "grade": "A+"}
students[1]. update(new_info) # 영희의 점수 업데이트

for s in students:
    # 'comment'라는 키가 없으면 "내용 없음"을 기본값으로 가져옴
    comment = s.get("comment", "내용 없음")
    print(f"학생: {s['name']}, 점수: {s['score']}, 코멘트: {comment}")

# 4.중급 챌린지: 딕셔너리 컴프리헨션
# 학새 이름을 키로, 점수를 값으로 하는 새로운 사전 만들기
score_map = {s["name"]: s["score"] for s in students}
print(f"\n한눈에 보는 성적표: {score_map}")
```

[요약]
- 복합 구조: 기차(리스트)의 칸마다 보물 상자(딕셔너리)가 실려 있는 것. 상자를 열면 그 안에 이름표와 보물이 들어있음
- .get(): "상자에 사탕 있어?" 라고 물어봤을 때, 없어도 울지 않고 "없으니까 0개라고 생각하자!" 라고 씩씩하게 말함
- 컴프리헨션: 여러개의 보물 중 "반짝이는 것만 골라서 새 바구니에 담아!"라고 한 줄로 명령하는 마법 주문

### 고급

고급 단계에서는 단순히 데이터를 담는 법을 너머. 파이썬이 내부적으로 이 데이터를 어떻게 관리하는지 그 '설계 원리'를 다룸. 특히 리스트와 딕셔너리를 다룰 때 가장 많이 실수하고 당황하는 부분을 정복

1. 고급 핵심 문법: 데이터의 내부 작동 원리
- 얕은 복사(Shallow Copy) vs 깊은 복사(Deep Copy) : 파이썬에서 리스트를 B = A 라고 복사하면, 데이터가 복사되는게 아니라 '똑같은 데이터를 가리키는 이름표'가 하나 더 생길뿐이다. 그래서 B를 바꾸면 A도 바뀌는 참사 발생. 그래서 중첩된 리스트나 딕셔너리를 완벽히 따로 복사하려면 copy.deepcopy()를 써야 함.
- 딕셔너리 뷰(view) 객체 : keys(),values(), items()는 리스트가 아니라 '뷰(view)'라는 특수한 객체임. 딕셔너리의 내용이 바뀌면 이 뷰 객체들도 실시간으로 같이 바뀜
- 리스트 내포(nested)와 메모리 참조 : 리스트 안에 또 리스트가 있을 때, 파이썬은 내부 리스트의 주소값만 저장함. 이 원리를 모르면 대규모 데이터를 다룰 때 예상치 못한 버그를 만날 수 있음.

```python
# [고급 - 복사(copy)의 비밀과 데이터 참조]

import copy

# 1.얕은 복사의 함정
original_list = [[1, 2], [3, 4]]
shallow_copy = original_list.copy()

# shallow_copy의 내부 리스트를 수정해봄
shallow_copy[0][0] = "바뀜!"

print("--- 얕은 복사 결과 ---")
print(f"원본: {original_list}") # 원본도 같이 바뀌어버림!!!
print(f"복사본: {shallow_copy}")

# 2.깊은 복사로 완벽하게 분리!
original_data = {"skills": ["Python", "AI"], "level": 1}
deep_copy = copy.deepcopy(original_data)

deep_copy["skills"].append("CUDA")

print("--- 깊은 복사 결과 ---")
print(f"원본 데이터: {original_data}")
print(f"깊은 복사본: {deep_copy}")

# 3.딕셔너리 컴프리센션 응용 (데이터 반전)
# 키와 값을 서로 뒤바꾸는 고급 기술
original_map = {"a": 1, "b": 2, "c": 3}
reversed_map = {v: k for k, v in original_map.items()}
print(f"\n키-값 반전 결과: {reversed_map}")
```

[요약]
- 얕은 복사: 사탕 바구니를 복사했는데, 알고 보니 바구니만 새로 사고 사탕은 원래 바구니에 끈으로 연결해둠.
- 깊은 복사: 사탕 바구니부터 그 안의 사탕까지 똑같이 하나 더 새로 만든 것과 같음. 이제 서로 아무 상관 없음.
- 데이터 반전: 이름표와 물건을 서로 바꿔치기하는 마법. "사과: 빨강"을 "빨강: 사과"로 바꿈.

#### 실습 과제
**과제 1: 전화번호부**

```python
# 빈 딕셔너리에서 시작해서
# 친구의 이름과 전화번호를 추가
# 그 후 친구의 전화번호를 검색하는 프로그램

phone_num = {}

phone_num["철수"] = "010-1234-5678"
print(phone_num)

new_info = {"영희": "010-1111-2222", "민수": "010-2222-3333"}
phone_num.update(new_info)
print(phone_num)

while True:
    search_name = input("찾을 이름을 입력하세요 : ")
    if search_name in phone_num:
        print(f"{search_name}의 연락처는 {phone_num[search_name]} 입니다.")
        break
    else:
        print("찾는 이름이 없습니다. 다시 입력해주세요")
```
