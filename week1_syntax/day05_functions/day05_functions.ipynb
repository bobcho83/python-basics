# 함수(Function)

1. 함수(function) : 함수는 '재료(입력)'를 넣으면 정해진 '순서'대로 요리해서 '완성된 접시(결과)'를 내보내는 자동 요리 기계와 같아요.
- 정의(def): 요리 기계를 만드는 것
- 호출: 요리 기계의 버튼을 눌러 실행하는 것
- 매개변수: 기계에 넣는 재료
- 반환(return): 기계가 툭 내 뱉는 완성품

```python
# 1.함수 정의와 호출(def, return) : def는 '이 기계를 지금부터 정의(define)하겠어!'라는 뜻
def 인사하기():
    print("안녕하세요! 반가워요!")

인사하기()
```

```python
# 2.매개변수와 기본값 : 재료를 줄 수도 있고, 안 주면 알아서 준비된 재료를 쓰게 할 수도 있어요.

def 선물하기(물건, 개수=1):
    print(f"{물건} {개수}개를 선물했습니다!")

선물하기("사과")
선물하기("포도", 3)
```

```python
# 3.반환값이 있는 함수
def add(a, b):
    result = a + b
    return result

sum_result = add(10, 14)
print(sum_result)

# 여러 반환값
def get_name_and_age():
    name = "철수"
    age = 25
    return name, age

name, age = get_name_and_age()
print(f"{name}의 나이는 {age} 입니다.")
```

```python
# [초급 - 함수 정의와 마법의 버튼]

# 1.매개변수와 기본값을 가진 함수
def make_juice(fruit="오렌지"):
    return f"상큼한 {fruit} 주스가 나왔습니다!"

print(make_juice("딸기"))
print(make_juice()) # 아무것도 안 넣으면 '오렌지' 주스가 나옴. 디볼트

# 2.여러값을 반환하는 함수(수학 비서)
def get_rectangle_info(width, height):
    area = width * height
    perimeter = (width + height) * 2
    return area, perimeter # 면적과 둘레를 동시에!

# 여러 개의 변수로 나눠서 받을 수 있음 (언팩킹)
area, peri = get_rectangle_info(10, 5)

print("\n--- 사각형 정보 ---")
print(f"가로 10, 세로 5인 사각형의 면적은 {area}, 둘레는 {peri}입니다.")

# 3.5일차 복습 퀴즈 함수
def check_score(score):
    if score >= 80:
        return "합격"
    else:
        return "재시험"

print(f"\n나의 결과: {check_score(95)}")
```

**중급 핵심 문법: 함수의 확장**

1.가변 매개변수 (*args, **kwargs) : 재료가 몇 개 들어올지 모를 때 사용.
- *args : 여러 개의 인자를 리스트(튜플) 형태로 한꺼번에 받음 (예: 점수들)
- **kwargs : 키워드와 값을 딕셔너리 형태로 한꺼번에 받음 (예: 유저 정보)

2.람다(Lambda) 함수 : 이름표도 붙이기 귀찮은 아주 간단한 계산은 한 줄짜리 익명 함수로 만듬. 코드가 굉장히 간결해짐.

3.변수의 범위(Scope) : 함수 안에서 만든 변수(Local)와 함수 밖에서 만든 변수(Global)의 차이를 이해해야 데이터가 꼬이지 않음.

```python
# [중급 - 유연한 함수와 람다 활용]

# 1. *args 활용: 몇 개가 들어와도 다 더해주는 함수
def calculate_sum(*numbers):
    # numbers는 튜플 형태로 들어옴
    total = sum(numbers)
    return total

print(f"3개 더하기: {calculate_sum(1, 2, 3)}")
print(f"5개 더하기: {calculate_sum(10, 20, 30, 40, 50)}")

# 2. **kwargs 활용: 유저 정보를 딕셔너리로 받기
def print_user_report(name, **details):
    print(f"\n--- {name}님의 리포트 ---")
    for key, value in details.items():
        print(f"- {key}: {value}")

# 이름 뒤에 원하는 정보를 마음껏 넣을 수 있음.
print_user_report("Bob", 나이=25, 목표="1일 1커밋", 주언어="Python")

# 3.Lambda함수 활용: 리스트 정렬하기
# 이름과 점수가 있는 리스트를 점수 기준으로 정렬
students = [("철수", 80), ("영희", 95), ("민수", 70)]

#점수(x[1])를 기준으로 정렬하라는 뜻
students.sort(key=lambda x: x[1], reverse=True)
print(f"\n성적 우수자 순위: {students}")
```

[요약]
1. *args: "뭐든지 다 담는 커다란 자루"로 사과를 1개 넣든 100개를 넣든 자루 하나에 다 담아주는 마법
2. **kwargs: "이름포가 붙은 종합 선물세트"로 "이건 나이", "이건 취미"라고 이름표를 붙여주면 함수가 딕셔너리로 예쁘게 정리

**고급 핵심 문법: 함수의 변신**
1. 일급 객체(First-class Object): 파이썬에서 함수는 변수로 취급. 함수를 다른 함수의 재료(인자)로 넣을 수도 있고, 함수 안에서 함수를 만들어 결과물(반환값)로 줄 수도 있음
2. 클로저(Closure): 함수 안에 또 다른 함수를 만드는 기법. 밖에 있는 함수가 종료되어도, 안쪽 함수가 그 당시의 데이터(상태)를 기억하고 있는 신기한 주머니 같은 기능
3. 데코레이터(Decorator): 이미 만들어진 함수를 수정하지 않고, 그 앞뒤에 새로운 기능을 '장식'하는 기술. @ 기호를 사용하며, 로그 기록이나 실행 시간 측정 등에 필수적

```python
# [고급 - 함수의 마법, 데코레이터]

# 1.데코레이터 정의(장식 만들기)
def start_end_decorator(func):
    def wrapper():
        print(f"--- '{func.__name__}' 함수 작업을 시작합니다 ---")
        func() # 실제 함수를 여기서 실행!
        print("--- 작업을 모두 마쳤습니다 ---\n")
    return wrapper

# 2.데코레이터 사용하기 (@기호 사용)
@start_end_decorator
def say_hello():
    print("안녕하세요, 밥(Bob)님! 파이썬 공부를 열심히 하시는군요!")

@start_end_decorator
def study_python():
    print("열심히 깃허브에 잔디를 심고 있습니다....🌱")

# 함수 호출
say_hello()
study_python()
```

[요약]
1. 일급 객체: 함수를 사탕처럼 가방(변수)에 넣거나 친구(다른 함수)에게 선물할 수 있다는 뜻
2. 클로저: 함수가 끝난 뒤에도 자기만의 비밀 일기장에 내용을 적어두고 나중에 꺼내 보는 것과 같음
3. 데코레이터: 이미 입고 있는 옷(함수)위에 멋진 외투(추가 기능)을 입혀주는 것. 옷 자체를 바꿀 필요 없이 겉모습(기능)만 화려하게 만드는 마법

```python
# 과제1: 두 수의 곱하기
# multiply 함수를 만들기
# 두 개의 숫자를 받아서 곱한 결과를 반환

def multiply(a, b):
    mul = a * b
    return mul

print(f"5 곱하기 3 = {multiply(5, 3)}입니다.")
print(multiply(10, 2))
```

```python
# 과제2: 성적 판정 함수
# grade 함수 만들기
# 점수를 받아서 학점을 반환

def grade(score):
    if score >= 90:
        point = "A"
    elif 90 > score >= 80:
        point = "B"
    elif 80 > score >= 70:
        point = "C"
    elif 70 > score >= 60:
        point = "D"
    else:
        point = "F"
    return point

print(grade(95))
print(grade(75))
```

```python
# 과제3: 리스트 역순 함수
# reverse_list 함수 만들기
# 리스트를 받아서 역순으로 정렬해서 반환

def reverse_list(a):
    new_list = list(reversed(a))
    return new_list

print(reverse_list([1, 2, 3, 4, 5]))
print(reverse_list([10, 9, 8, 7, 6]))
```
