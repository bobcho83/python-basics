**7일차: 미니 프로젝트 - 스마트 계산기 만들기**
- 단순히 계산만 하는 게 아니라, 프로그램이 멈추지 않도록 오류를 방지(예외 처리)하는 고급스러운 설계

[핵심 포인트]
1. 입력 받기: input()으로 받은 문자열을 float()나 int()로 바꿔줌. (소수점 계산도 가능하게 float를 추천)
2. 연산 선택: if-elif-else 문으로 연산자를 판별
3. 예외처리: try-except문을 사용해서 0으로 나누는 상황이나 숫자가 아닌 글자를 입력하는 상황을 방버

```python
# [day 7: 미니 프로젝트 - 예외 처리를 포함한 스마트 계산기]

def calculator():
    print("--- 밥(Bob)의 스마트 계산기 시작 ---")

    try:
        # 1.숫자 입력받기
        num1 = float(input("첫 번째 숫자를 입력하세요: "))
        num2 = float(input("두 번째 숫자를 입력하세요: "))

        # 2.연산자 선택받기
        operator = input("어떤 연산을 할까요? (+, -, *, /): ")

        # 3.계산 및 출력(조건문 활용)
        if operator == '+':
            result = num1 + num2
        elif operator == '-':
            result = num1 - num2
        elif operator == '*':
            result = num1 * num2
        elif operator == '/':
            # 4.0으로 나누기 오류 처리 (ZeroDivisionError)
            if num2 == 0:
                return "오류: 0으로 나눌 수 없습니다."
            result = num1 / num2
        else:
            return "오류: 지원하지 않는 계산입니다."

        return f"결과: {num1} {operator} {num2} = {result}"
    
    except ValueError:
        # 숫자가 아닌 글자를 입력했을 때 발생하는 오류 처리
        return "오류: 숫자만 입력해주세요!"

# 프로그램 실행
print(calculator())
```

[핵심 요약]
1. try - except: '안전벨트' 차가 덩컹(에러 발생)거려도 다치지(프로그램 종료)않게 꽉 잡아주는 역할
2. float(): int는 정수만 담지만, float는 소수점(3.14 같은)도 담을 수 있는 '넉넉한 상자'
3. ZeroDivisionError: 수학에서 어떤 숫자를 0으로 나누는 건 불가능. 그래서 컴퓨터가 당황하기 전에 "0으로 나누면 안 돼!"라고 미리 규칙을 정해주는 것.

**미니 프로젝트 - 스마트 할 일 관리 (To-Do List)**
- 핵심 문법 : list.append(), list.pop(), enumerate(), while

```python
# 1.할 일 목록 프로젝트
todo_list = []

while True:
    print(f"현재 할 일({len(todo_list)}개): {todo_list}")
    menu = input("1.추가 2.삭제 3.종료")

    if menu == "1":
        job = input("할 일을 입력하세요: ")
        todo_list.append(job)
    elif menu == "2":
        # enumerate는 번호를 매겨주는 고마운 함수!
        for idx, item in enumerate(todo_list):
            print(f"{idx}: {item}")
        delete_idx = int(input("삭제할 번호를 선택하세요: "))
        todo_list.pop(delete_idx)
    elif menu == "3":
        print("할 일 관리를 종료합니다.")
        break
```

**미니 프로젝트 - 마케팅 성과 분석기(Daily Report)**
- 핵심 문법: sum(), len(), for문 필터링, 리스트 연산

```python
# 2.마케팅 데이터 분석 프로젝트
campaign_clicks = [120, 340, 50, 210, 400] # 샘플 데이터
high_performance = []

# 평균 계산
avg_clicks = sum(campaign_clicks) / len(campaign_clicks)

# 반복문으로 평균보다 잘 나온 성과만 뽑기
for click in campaign_clicks:
    if click > avg_clicks:
        high_performance.append(click)

print(f"--- 오늘의 마케팅 리포트 ---")
print(f"전체 평균 클릭 수: {avg_clicks}")
print(f"평균 돌파 캠페인 수: {len(high_performance)}개")
print(f"최고 기록: {max(campaign_clicks)}")
```

**미니 프로젝트 - 숫자 맞히기 & 명예의 전당
- 핵심 문법: while True, if-elif, list.sort(), count 변수

```python
# 3. 업다운 게임 & 랭킹 프로젝트
import random
hall_of_fame = [] # 시도 횟수를 저장하는 명예의 전당 리스트

while True:
    target = random.randint(1, 100) # 1~100 사이 랜덤 숫자
    count = 0
    print("\n--- 새로운 게임 시작 (1~100) ---")

    while True:
        guess = int(input("숫자를 맞춰보세요: "))
        count += 1

        if guess == target:
            print(f"정답! {count}번 만에 맞히셨네요.")
            hall_of_fame.append(count)
            break
        elif guess < target:
            print("UP!")
        else:
            print("DOWN!")

    # 명예의 전당 정렬 (시도 횟수가 적은 순서)
    hall_of_fame.sort()
    print(f"현재 1등 기록: {hall_of_fame[0]}회")

    retry = input("다시 하시겠습니다? (y/n): ")
    if retry.lower() != 'y': break
```

[요약]
- 할 일 관리: append로 기차 칸을 붙이고, pop으로 기차 칸을 떼어내는 연습
- 성과 분석: 흩어져 있는 숫자들을 for 라는 바구니에 담아 검사하고, 우수한 녀석들만 따로 모으는 법
- 랭킹 게임: sort()를 써서 누가 가장 적게 시도했는지 순서대로 줄을 세우는 방법
