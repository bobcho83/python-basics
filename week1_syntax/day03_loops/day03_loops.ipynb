# 반복문 for / while

반복문은 컴퓨터가 지치지 않고 똑같은 일을 수만 번 대신 하게 만드는 기술

1. 반복문 : 컴퓨터에게 "사과 먹어"라고 100번 말하기는 힘들때 반복문을 쓴다.
- for 문 : "정해진 횟수만큼" 또는 "바구니(리스트)에 담긴 알맹이 수만큼" 반복. (주로 끝이 정해져 있을 때 사용)
- while 문 " "~하는 동안은 계속해!"라고 말하는 것. 조건이 맞으면 무한히 반복할 수 있도 있음. (주로 특정 상황이 끝날 때까지)

### 초급

```python
# 1)for문과 range
# range(5)라고 하면 0부터 4까지, 총 5개의 숫자를 만들어줌.

for i in range(5):
    print(i) # 0, 1, 2, 3, 4
print("="*30)

fruits = ["사과", "배", "포도"]
for f in fruits:
    print(f) # 사과, 배, 포도 차례대로 출력
```

```python
# [초급 반복문 - 구구단과 리스트 순회]

print("---미션 1: 구구단 2단 만들기 (for문)---")
# 1부터 9까지 곱해야 하니까 range(1, 10) 사용
for i in range(1, 10):
    print(f"2 * {i} = {2*i}")
print("="*30)
print("\n")

print("---미션 2: 출석부 부르기 (리스트 순회)---")
students = ["철수", "영희", "민수", "Bob"]
for name in students:
    print(f"{name} 학생! 왔나요?")
print("="*30)
print("\n")

print("---미션 3: 1부터 5까지 숫자 세기 (while)---")
number = 1
while number <= 5:
    print(f"현재 숫자는 {number} 입니다.")
    number = number + 1 # 숫자를 1씩 키워줘야 끝이 남.
```

```python
# 구구단 변형 - input으로 입력 받아 구구단 하기!

while True:
    number = int(input("2~9까지만 입력해주세요!"))

    if number > 1 and number < 10:
        print(f"{number}단 시작!!")
        for i in range(1,10):
            print(f"{number} * {i} = {number * i}")
        break # 구구단이 성공했을 때 while 문에서 빠져나갈 수 있는 장치 브레이크!!
    else:
        print("2~9까지만 입력해주세요!!")
print("="*30)
print("코드 심플하게 줄이기!")

while True:
    num = int(input("2~9까지만 입력해주세요!"))

    if 1 < num < 10:
        print(f"{num}단 시작!!")
        for i in range(1,10):
            print(f"{num} * {i} = {num * i}")
        break
    print("2~9까지만 입력해주세요!!") #else를 생략할 수 있다.
```

### 중급
중급 반복문의 핵심은 '반복문 안에 반복문 넣기(중첩 루프)'와 '반복문을 더 똑똑하게 제어하기'이다.

1. 중급 핵심 문법 : 반복문의 진화
- 이중 for문 (Nested Loops) : 반복문 안에 또 다른 반복문을 넣는 방식. 시계의 '분침이 60번 돌아야 시침이 1번 움직이는 것'과 같은 원리. 구구단 전체 (2~9단)을 만들 때 필수!
- break와 continue : break > 당장 멈춰! (반복문 탈출), continue > '이번 판은 건너뛰고 다음 판으로 가자!' (아래 코드를 무시하고 다음 반복으로 넘어감)
- enumerate() : 리스트를 순회할 때 '지금이 몇 번째(인덱스)인지'와 '값'을 동시에 알려주는 아주 친절한 함수

```python
# [중급 반복문 - 중첨 루프와 제어문]

print("---미션 1: 구구단 전체 출력 (이중 for문)---")
# 바깥쪽 for문은 '단' (2단부터 9단까지)
for dan in range(2, 10):
    print(f"{dan}단 시작!")
    for i in range(1, 10):
        print(f"{dan} * {i} = {dan * i}")
```

```python
print("---미션 2: 짝수단만 출력하기! (continue 활용)---")
for dan in range(2, 10):
    if dan % 2 != 0: # 홀수단이라면
        continue
    else:
        print(f"{dan}단 시작!")
        for i in range(1, 10):
            print(f"{dan} * {i} = {dan * i}")
```

```python
print("---미션 3: 인덱스와 함께 출력 (enumerate)---")
tasks = ["메일 확인", "파이썬 공부", "운동", "깃허브 업로드"]
for index, task in enumerate(tasks, start=1):
    print(f"{index}순위 할 일: {task}")
```

### 고급
고급 반복문의 핵심은 '얼마나 짧고 강력하게 쓰는가'와 '메모리를 얼마나 아끼는가'에 있다. 파이썬의 정수라고 불리는 기법을 남긴다.

1. 고급 핵심 문법 : 파이썬답게 반복하기
- 리스트 컴프리헨션(List Comprehension) : for문과 if문을 대괄호[]안에 한 줄로 압축하는 기술이다. 코드가 비약적으로 짧아지고 읽기 편해짐.
- zip() 함수 : 두 개 이상의 리스트를 병렬로 묶어서 동시에 반복문을 돌릴 때 사용. '이름 리스트'와 '점수 리스트'를 한꺼번에 처리할 때 유용.
- 제너레이터 표현식(Generator Expression) : 데이터가 수백만 개일 때 리스트로 만들면 컴퓨터가 멈출 수 있다. 제너레이터는 '필요할 때만 하나씩' 값을 생성해서 메모리를 거의 사용하지 않는 영리한 방법.

```python
# [고급 반복문 - 컴프리헨션과 호율적 병력 처리]

# 1.리스트 컴프리헨션: 1부터 10까지 숫자 중 '짝수'만 제곱해서 담기
# (표현식 for 항목 in 이터러블 if 조건)
even_squares = [x**2 for x in range(1, 11) if x % 2 == 0]
print(f"짝수 제곱 리스트: {even_squares}")

# 2.zip()을 활용한 병렬 순회
# 두 개의 리스트를 지퍼(zip)처럼 딱 맞물려 하나씩 꺼냄.
subjects = ["python", "AI", "Marketing"]
scores = [100, 95, 85]

print("\n--- 과목별 성적표 ---")
for sub, score in zip(subjects, scores):
    print(f"{sub} 과목 점수: {score}점")

# 3.제너레이터 표현식 (메모리 절약형 반복)
# 대괄호[] 대신 소괄호 ()를 사용.
# 만약 range가 10억이라면, 리스트는 터지지만 제너레이터는 멀쩡!!!
huge_range = (n * 10 for n in range(1000000))
print(f"\n제너레이터 객체 상태: {huge_range}")
print(f"첫 번째 값 꺼내기: {next(huge_range)}")
print(f"두 번째 값 꺼내기: {next(huge_range)}")
```

[고급 문법 요약]
1. 컴프리헨션 : 여러 줄에 걸쳐 쓰던 일기를 딱 한 줄로 요약해서 쓰는 마법의 문장.
2. zip() : 운동화 끈을 묶듯이, 왼쪽 리스트와 오른쪽 리스트를 한 쌍으로 묶어서 같이 데려감.
3. 제너레이터 : 과자를 미리 다 까놓는 게 아니라, '한 개만 줘!' 라고 할 때마다 봉지에서 하나씩 꺼내주는 인색하지만 똑똑한 친구! 메모리 절약!!!
