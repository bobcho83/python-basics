2. Series (시리즈): 판다스의 가장 기본 단위로, **'인덱스(라벨)'가 붙은 1차원 배열**임. 파이썬의 리스트와 딕셔너리의 장점을 합쳐놓은 형태
3. 인덱스와 데이터 접근: 데이터에 0, 1, 2 같은 숫자 번호뿐만 아니라 '날짜', '이름'같은 나만의 라벨을 붙일 수 있고, 이 라벨로 데이터를 쏙쏙 골라낼 수 있음

```python
# [Pandas 기초 - Series 생성과 분석]

import pandas as pd # 관례적으로 pd라는 별명을 사용

# 1.Series 생성 (데이터와 인덱스 설정)
# 일주일간의 광고 클릭 수 데이터
data = [120, 150, 90, 200, 180, 130, 210]
days = ['월', '화', '수', '목', '금', '토', '일']

clicks = pd.Series(data, index=days)

print("--- 주간 클릭 수 Series ---")
print(clicks)

# 2.데이터 접근 (인덱싱)
print(f"\n수요일 클릭 수: {clicks['수']}") # 라벨로 접근
print(f"첫 번째 데이터: {clicks[0]}") # 순서(숫자)로 접근

# 3.기본 메서드 활용 (통계 분석)
print("\n--- 데이터 요약 ---")
print(f"총 클릭 수 (sum): {clicks.sum()}")
print(f"평균 클릭 수 (mean): {clicks.mean():.2f}")
print(f"최대 클릭 수 (max): {clicks.max()}")
print(f"최소 클릭 수 (min): {clicks.min()}")

# 4.필터링 (조건에 맞는 데이터 찾기)
print("\n[클릭 수가 150회 이상인 날]")
print(clicks[clicks >= 150])
```

**[핵심 요약]**
1. pandas: 엄청나게 똑똑한 '엑셀 로봇'. 코드로 명령만 내리면 복잡한 계산을 순식간에 해줌
2. Series: 엑셀 시트에서 딱 '세로 한 줄'을 떼어낸 것
3. 인덱스: 상자에 붙인 '이름표'임. "3번째 상자 가져와!"라고 하는 대신 "월요일 상자 가져와!"라고 할 수 있게 해줌
4. 기본 메서드: 로봇에게 시키는 '간단한 심부름'임. "다 더해줘(sum)", "평균 내줘(mean)"라고 말 하면 끝!

#### 중급 핵심 문법: 데이터프레임(DataFrame)기초
1. DataFrame이란?: 여러 개의 시리즈(Series)가 나란히 붙어 있는 2차원 표 구조임. 행(Row)과 열(Column)이 있고, 각각 이름표(인덱스, 컬럼명)가 붙어 있음.
2. 딕셔너리로 생성하기: 가장 직관적인 방법은 파이썬 딕셔너리를 이용하는 것. Key는 열 이름(Column)이 되고, Value는 그 열에 들어갈 데이터 리스트가 됨
3. 데이터 확인하기(head, tail, info): 데이터가 너무 많을 때 앞부분만 보거나(head), 전체적인 요약 정보(info)를 확인할 수 있음

```python
# [Pandas 중급 - DataFrame 생성과 탐색]

import pandas as pd

# 1.딕셔너리를 이용해 데이터프레임 만들기
data = {
    "채널": ["Google", "Facebook", "Instagram", "YouTube", "X"],
    "노출수": [10000, 8500, 12000, 15000, 5000],
    "클릭수": [450, 320, 580, 700, 150],
    "비용": [150000, 120000, 180000, 250000, 50000]
}

df = pd.DataFrame(data)

# 2.데이터 확인하기
print("--- 마케팅 성과 데이터프레임 ----")
print(df)

# 3.특정 열(Column)만 선택하기 (Series 반환)
print("\n[채널명 리스트]")
print(df["채널"])

# 4.새로운 열 계산해서 추가하기 (파생 변수)
# 클릭률(CTR) = 클릭수 / 노출수 * 100
df["CTR(%)"] = (df["클릭수"] / df["노출수"]) * 100

print("\n--- CTR이 추가된 데이터 ---")
print(df.head(3)) # 상단 3개 데이터만 확인

# 5.데이터 요약 통계 확인
print("\n--- 수치 데이터 요약 ---")
print(df.describe()) # 평균, 표준편차, 최솟값 등을 한눈에!
```

**[핵심 요약]**
1. DataFrame: 시리즈들이 모여서 만들어진 '커다란 바둑판 표'임. 엑셀 화면 그 자체라고 생각하면 됨
2. 딕셔너리로 만들기: "제목: [내용들]" 형식으로 정리해서 판다스에게 주면, 예쁘게 줄을 맞춰서 표를 그려줌
3. df["새로운이름"]: 기존에 있던 숫자들을 이용해서 '새로운 칸'을 만드는 것. 노출수랑 클릭수를 나눠서 '얼마나 인기가 많은지'를 계산하는 칸을 뚝딱 만들 수 있음
4. describe(): 우리 반 친구들의 수학 점수를 다 더하고 평균 내고 1등 점수 를 찾는 과정을 단 한마디에 '성적표 요약'으로 보여주는 것

#### 고급 핵심 문법: 데이터 자유자재로 다루기
1. 조건 필터링(Boolean Indexing): df[조건] 형식을 사용하여 특정 기준을 만족하는 행만 추출. 엑셀의 '필터' 기능과 똑같음
2. .loc와 .iloc
- .loc: 이름을 보고 찾아감 (label-based)
- .iloc: 번호(순서)를 보고 찾아감 (Integer-based)
3. 정렬하기 (sort_values): 특정 열의 값을 기준으로 오름차순(작은 순) 또는 내림차순(큰 순)으로 줄을 세움

```python
# [pandas 고급 - 데이터 필터링과 정렬]

import pandas as pd

data = {
    "채널": ["Google", "Facebook", "Instagram", "YouTube", "X"],
    "노출수": [10000, 8500, 12000, 15000, 5000],
    "클릭수": [450, 320, 580, 700, 150],
    "비용": [150000, 120000, 180000, 250000, 50000]
}

df = pd.DataFrame(data)
df["CTR(%)"] = (df["클릭수"] / df["노출수"]) * 100

# 1.조건 필터링: CTR이 4% 이상인 채널만 추출
high_ctr = df[df["CTR(%)"] >= 4.0]
print("--- CTR 4% 이상 채널 ---")
print(high_ctr)

# 2.정렬하기: 클릭수 기준 내림차순 (큰 것부터)
sorted_df = df.sort_values(by="클릭수", ascending=False)
print("\n--- 클릭수 기준 정렬 ---")
print(sorted_df)

# 3. .loc 활용: 특정 행과 열 선택
# 인덱스 0부터 2까지, '채널'과 'CTR(%)' 열만 보고 싶을 때
print("\n--- 특정 범위 선택 (.loc) ---")
print(df.loc[0:2, ["채널", "CTR(%)"]])

# 4.다중 조건: 노출수가 10000 이상이면서 비용은 20만원 이하인 경우
filtered_ads = df[(df["노출수"] >= 10000) & (df["비용"] <= 200000)]
print("\n--- 가성비 채널 (노출 1만 이상 & 비용 20만 이하) ---")
print(filtered_ads)
```

**[핵심 요약]**
1. 조건 필터링: 학교 급식실에서 '키 150cm 이상만 줄 서!'라고 하는 것과 같음. 조건에 안 맞는 데이터는 잠시 숨기고, 맞는 데이터만 보여주는 것
2. .loc: '이름표'로 찾는 것. "철수네 집이랑 영희네 집 정보 보여줘!"라고 하는 것
3. .iloc: '번호'로 찾는 것. "앞에서 1번째부터 3번째 집까지 보여줘!"라고 순서로 말하는 것
4. 정렬(sort_values): 키 순서대로 줄을 세오는 것. 내림차순은 키가 제일 큰 형이 맨 앞에서 서는 방식
