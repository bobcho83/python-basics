# DataFrame 생성 및 조회

#### DataFrame 구조와 생성 핵심
1. DataFrame이란? 행(Row)과 열(Column)로 구성된 2차원 표 형태의 데이터 구조입니다.
2. 데이터프레임 만들기: 가장 많이 쓰이는 방법은 딕셔너리를 이용하는 것. 딕셔너리의 Key가 열(Column)의 이름이 되고, 리스트 데이터가 그 열의 내용이 됨
3. 데이터 접근 (조회)
- 열 접근: df['컬럼명']을 사용해 특정 열만 뽑아냄
- 행 접근: df.loc 이름(라벨)을 기준으로 행을 찾음 / df.iloc 컴퓨터가 매기는 순서(인덱스 번호)를 기준으로 행을 찾음

```python
# [Pandas DataFrame 생성 및 조회 기초]

import pandas as pd

# 1.DataFrame 만들기 (딕셔너리 방식)
data = {
    "캠페인명": ["여름세일", "신규가입", "리마케팅", "브랜드홍보"],
    "예산": [500000, 300000, 200000, 1000000],
    "클릭수": [1200, 800, 450, 2100],
    "지역": ["서울", "경기", "부산", "전국"]
}

df = pd.DataFrame(data)

# 2.데이터 구조 확인
print("--- 전체 데이터프레임 ---")
print(df)
print(f"\n데이터 크기 (행, 열): {df.shape}")

# 3.열(Column)에 접근하기
print("\n[예산 컬럼 정보]")
print(df["예산"])

# 4.행(Row)에 접근하기
# loc: 인덱스 라벨 '0'인 행 (보통 숫자 인덱스일 땐 번호와 같음)
print("\n[첫 번째 행 상세 정보 (loc)]")
print(df.loc[0])

# iloc: 2번째(인덱스 1)부터 3번째(인덱스 2) 행까지 슬라이싱
print("\n[두 번째부터 세 번째 행까지 (iloc)]")
print(df.iloc[1:3])

# 5.특정 행과 열을 동시에 접근
print(f"\n세 번째 캠페인의 클릭수: {df.loc[2, '클릭수']}")
```

**[핵심 요약]**
1. DataFrame: 아주 커다란 '바둑판 일기장'임. 가로줄(행)에는 날짜별 기록을 쓰고, 세로줄(열)에는 날짜, 날씨, 기분 같은 제목을 붙여 정리하는 것
2. 행(Row): 일기장에서 '한 줄의 기록'임. "오늘의 일기" 하나가 행 하나가 됨
3. 열(Column): 일기장의 '항목'임. "기분"이라는 칸만 쭉 보면 이번 주 내 기분이 어땠는지 한눈에 알 수 있음
4. loc vs iloc
- loc는 "주소"로 찾는 것. "서울 사시는 분 손 들어보세요!"라고 이름을 부르는 것
- iloc는 "번호"로 찾는 것. "앞에서 2번째 앉은 사람 나와보세요!"라고 순서를 부르는 것

#### 중급 핵심 문법: 데이터 자유자재로 다루기
1. 조건 필터링 (Boolean Indexing): df[조건] 형식을 사용하여 특정 기준을 만족하는 행만 추출. 엑셀의 '필터' 기능과 똑같음
2. 다중 조건 필터링: 여러 개의 조건을 동시에 걸 수 있음
- & (And): 두 조건 모두 만족
- | (or): 둘 중 하나라도 만족
3. 데이터 정렬 (sort_values): 특정 열의 값을 기준으로 오름차순(작은 순)또는 내림차순(큰 순)으로 순서를 바꿈

```python
# [Pandas 중급 - 데이터 필터링과 정렬]

import pandas as pd

data = {
    "캠페인명": ["여름세일", "신규가입", "리마케팅", "브랜드홍보"],
    "예산": [500000, 300000, 200000, 1000000],
    "클릭수": [1200, 800, 450, 2100],
    "지역": ["서울", "경기", "부산", "전국"]
}
df = pd.DataFrame(data)

# 1.새로운 열 계산 (클릭당 비용 등 파생 변수 생성)
df["클릭당_비용"] = df["예산"] / df["클릭수"]

# 2.조건 필터링: 클릭수가 1000회 이상인 캠페인만 추출
high_clicks = df[df["클릭수"] >= 1000]
print("--- 클릭수 1000회 이상 캠페인 ---")
print(high_clicks)

# 3.다중 조건: 지역이 '서울'이면서 예산이 20만원 이상인 캠페인
seoul_expensive = df[(df["지역"] == "서울") & (df["예산"] >= 200000)]
print(seoul_expensive)

# 4.데이터 정렬: 클릭당 비용이 낮은 순서대로(오름차순) 정렬
sorted_df = df.sort_values(by="클릭당_비용", ascending=True)
print(sorted_df)
```

**[핵심 요약]**
1. 조건 필터링: 학교 급식실에서 '안경 쓴 사람만 줄 서!'라고 하는 것과 같음. 수많은 데이터 중에서 내가 원하는 특징을 가진 데이터만 따로 모으는 것
2. 다중 조건(&): "안경도 쓰고, 파란색 옷도 입은 사람!"처럼 두 가지 조건을 다 지켜야 할 때 사용하는 아주 꼼꼼한 필터
3. 정렬(sort_values): 키 순서대로 줄을 세우는 것. 오름차순(True)은 작은 사람부터, 내림차순(False)은 큰 사람부터 서는 방식

#### 고급 핵심 문법: 데이터 집약과 연결
1. 데이터 병합(Merge): 두 개의 데이터 프레임에 공통된 열(Key)이 있다면, 이를 기준으로 두 표를 하나로 합칠 수 있음. 엑셀의 VLOOKUP 기능과 비슷
2. 그룹화(Groupby): 특정 열을 기준으로 데이터를 그룹으로 묶고, 각 그룹의 합계(sum), 평균(mean), 개수(count) 등을 계산함

```python
# [Pandas 고급 - 데이터 병합과 그룹화]

import pandas as pd

# 1.두 개의 데이터 프레임 준비
df_campaign = pd.DataFrame({
    "캠페인명": ["여름세일", "신규가입", "리마케팅", "반짝특가"],
    "클릭수": [1200, 800, 450, 300],
    "지역코드": ["A", "B", "A", "C"]
})

df_fegion = pd.DataFrame({
    "지역코드": ["A", "B", "C"],
    "지역명": ["서울", "경기", "부산"]
})

# 2.데이터 병합 (Merge): '지역코드'를 기준으로 합치기
merged_df = pd.merge(df_campaign, df_fegion, on="지역코드")
print("--- 병합 된 데이터프레임 ---")
print(merged_df)

# 3.그룹화 (Groupby): 지역명별로 클릭수 합계 구하기
region_summary = merged_df.groupby("지역명")["클릭수"].sum()
print("\n--- 지역별 총 클릭수 요약 ---")
print(region_summary)

# 4.여러 개의 통계 한꺼번에 보기 (agg)
detail_summary = merged_df.groupby("지역명")["클릭수"].agg(["sum", "mean", "count"])
print("\n--- 지역별 상세 통계 (합계, 평균, 개수) ---")
print(detail_summary)
```

**[핵심 요약]**
1. 데이터 병합(Merge): 두 개의 '퍼즐 조각'을 맞추는 것과 같음. 한 조각에는 '상품 이름'이 있고 다른 조각에는 '상품 가격'이 있을 때, '상품 번호'라는 공통된 모양을 보고 두 조각을 하나로 합치는 것
2. 그룹화(groupby): 흩어져 있는 사탕들을 '색깔별로 바구니에 담는 것'과 같음. 빨간 바구니(서울), 파란 바구니(경기)에 사탕을 나눠 담은 뒤, 각 바구니에 사탕이 몇 개 있는지 세어보는 과정
