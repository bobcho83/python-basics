# 데이터 필터링과 정렬
- 3주차 핵심인 판다스(pandas)를 활용해, 수만 개의 데이터 중 내가 원하는 정보만 쏙쏙 골라내는 법을 학습

#### 18일차 핵심 개념 정리
1. 조건 필터링: df[df['컬럼'] >. 값]처럼 대괄호 안에 조건을 넣어 true인 데이터만 가져옴
2. 여러 조건: &(그리고) |(또는)를 사용. 이때 각 조건은 반드시 소괄호 ()로 감싸야 에러가 나지 않음
3. 정렬(sort_values): 특정 열을 기준으로 데이터를 줄 세움. ascending=False를 쓰면 내림차순(큰 것부터)이 됨
4. 범위 선택(loc, iloc)
- loc: 이름(라벨)으로 범위를 선택 (끝 번호 포함)
- iloc: 위치(인덱스 번호)로 범위를 선택 (끝 번호 제외)
5. isin 메서드: 리스트 안에 포함된 값들이 있는지 한꺼번에 검사할 때 매우 편리

```python
# [Pandas 데이터 필터링 및 정렬 마스터하기]

import pandas as pd

# 1.샘플 데이터 생성
data = {
    '캠페인': ['여름세일', '신규가입', '리마케팅', '브랜드홍보', '반짝특가'],
    '채널': ['Google', 'Facebook', 'Instagram', 'Google', 'YouTube'],
    '클릭수': [1200, 800, 450, 2100, 300],
    '비용': [500000, 300000, 200000, 1000000, 150000],
    '지역': ['서울', '경기', '부산', '서울', '전국']
}
df = pd.DataFrame(data)

# 1-2.기본 필터링: 클릭수가 1000회 이상인 캠페인
high_clicks = df[df['클릭수'] >= 1000]
print("--- 클릭수 1000회 이상 ---")
print(high_clicks)

# 2.여러 조건: 지역이 '서울'이면서 비용이 80만원 이하인 캠페인
# (반드시 소괄호를 써야 함)
seoul_efficient = df[(df['지역'] == '서울') & (df['비용'] <= 800000)]
print("\n--- 서울 지역 가성비 캠페인 ---")
print(seoul_efficient)

# 3.정렬: 클릭수 기준 내림차순 정렬
sorted_df = df.sort_values(by='클릭수', ascending=False)
print("\n--- 클릭수 높은 순 정렬 ---")
print(sorted_df)

# 4.범위 선택 (loc, iloc)
# loc는 '브랜드홍보' 행을 이름으로 찾을 때 유용하지만, 여기선 인덱스 활용
print("\n--- 0번부터 2번 행까지, 캠페인과 클릭수만 보기 (loc) ---")
print(df.loc[0:2], ['캠페인', '클릭수'])

# 5.isin 메서드: 특정 채널(Google, Instagram)만 포함된 데이터 찾기
target_channels = df[df['채널'].isin(['Google', 'Instagram'])]
print("\n--- 주요 채널 성과 조회 (isin) ---")
print(target_channels)
```

**[핵심 요약]**
1. 필터링: 학교 급식실에서 "안경 쓴 친구들만 이쪽으로 줄 서!"라고 분리하는 것과 같음. 내가 정한 규칙에 맞는 데이터만 따로 모으는 것
2. 여러 조건: "안경도 쓰고, 파란색 운동화도 신은 친구!"처럼 두 가지 조건을 모두 지켜야 할 때 사용하는 아주 꼼꼼한 필터
3. 정렬: 키 순서대로 줄을 세우는 것. 내림차순은 키가 제일 큰 친구가 맨 앞에 서는 방식
4. isin: "우리 반 명단에 이름이 있는 사람 손들어!"라고 물어보는 것과 같음. 리스트에 적힌 이름이 있는지 한 번애 확인해주는 것

#### [중급 핵심 문법: 데이터 요약과 가공]
1. 그룹화(groupby): 특정 기준(예:채널, 지역)으로 데이터를 묶음. 엑셀의 피벗 테이블과 같은 역할
2. 집계(agg): 그룹화한 데이터에 대해 합계(sum), 평균(mean), 개수(count)등 여러 통계를 한 꺼번에 구함
3. 파생 변수 생성: 기존 컬럼들을 계산하여 새로운 컬럼(예: 클릭률 CTR)을 만듬
4. 데이터 치환(replace, map): 특정 값을 다른 값으로 바꾸거나(예: 'Seoul' -> '서울'), 딕겨너리를 이용해 일괄 변경함

```python
# [Pandas 중급 - 데이터 그룹화 및 파생 변수 생성]

import pandas as pd

# 1.샘플 데이터 생성
data = {
    '채널': ['Google', 'Facebook', 'Instagram', 'Google', 'Facebook', 'Instagram'],
    '노출수': [10000, 8000, 12000, 15000, 9000, 11000],
    '클릭수': [450, 320, 580, 700, 400, 550],
    '비용': [150000, 120000, 180000, 250000, 140000, 170000]
}
df = pd.DataFrame(data)

# 2.파생 변수 생성: 클릭률(CTR) 계산
# 클릭률 = (클릭수 / 노출수) * 100
df['CTR(%)'] = (df['클릭수'] / df['노출수']) * 100

# 3.그룹화 (Groupby): 채널별로 데이터 묶기
# 채널별 총 클릭수와 평균 비용을 한눈에 확인
channel_summaary = df.groupby('채널').agg({
    '클릭수': 'sum',
    '비용': 'mean',
    'CTR(%)': 'mean'
})

print("---- 채널별 성과 요약 ---")
print(channel_summaary)

# 4.데이터 정렬: 평균 CTR이 높은 순서대로 정렬
sorted_summary = channel_summaary.sort_values(by='CTR(%)', ascending=False)
print("\n--- 가성비 채널 순위 (CTR 기준) ---")
print(sorted_summary)

# 5. 데이터 값 치환 (Map)
# 채널 이름을 한글로 보기 좋게 변경
name_map = {'Google': '구글', 'Facebook': '페이스북', 'Instagram': '인스타'}
df['채널_한글'] = df['채널'].map(name_map)

print("\n--- 이름 치환 결과 ---")
print(df[['채널', '채널_한글']].head())
```

#### [고급 핵심 문법: 입체적 데이터 분석]
1. 복합 필터링(query): df.query()를 사용하면 마치 SQL 문장처럼 직관적으로 데이터를 필터링 할 수 있음. 조건이 많아질 때 코드가 훨씬 깔끔해짐
2. 시계열 변환(to_datetime): 문자열로 된 날짜를 파이썬이 계산할 수 있는 '진짜 날짜'로 바꿈
3. 날짜 기반 추출: 날짜에서 연, 월, 일, 요일을 쏙쏙 뽑아내서 요일별/월별 리포트를 만듬
4. 조건부 값 할당 (np.where): 엑셀의 IF 함수처럼 특정 조건에 맞춰 결과값을 바로 생성

```python
# [Pandas 고급 - 시계열 데이터 분석 및 복합 필터링]

import pandas as pd
import numpy as np

# 1.시계열 포함 샘플 데이터 생성
data = {
    '날짜': ['2026-05-01', '2026-05-15', '2026-06-01', '2026-06-20', '2026-07-05'],
    '채널': ['Google', 'Facebook', 'Google', 'Instagram', 'YouTube'],
    '클릭수': [1200, 800, 2100, 450, 3000],
    '비용': [500000, 300000, 800000, 200000, 1200000]
}
df = pd.DataFrame(data)

# 2.시계열 데이터 처리 (글자를 날짜로!)
df['날짜'] = pd.to_datetime(df['날짜'])
df['월'] = df['날짜'].dt.month
df['요일'] = df['날짜'].dt.day_name()

# 3.np.where를 활용한 효율성 등급 부여 (IF문 스타일)
# 클릭수가 1000이상이고 비용이 60만원 이하인 경우 '우수', 아니면 '일반'
df['효율등급'] = np.where((df['클릭수'] >= 1000) & (df['비용'] <= 600000), '우수', '일반')

# 4.복합 필터링(Query 사용)
# 5월 데이터이면서 클릭수가 1000이 넘는 데이터만 추출
may_top_ads = df.query("월 == 5 and 클릭수 == 1000")
print("--- 5월 고효율 광고 추출 ---")
print(may_top_ads)

# 5.피벗 테이블 (Pivot Table): 월별/채널별 총 비용 요약
monthly_pivot = df.pivot_table(index='월', columns='채널', values='비용', aggfunc='sum', fill_value=0)
print("\n--- 월별/채널별 지출 요약 ---")
print(monthly_pivot)
```
