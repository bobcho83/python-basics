- pip install matplotlib : 파이썬의 가장 기본적이고 강력한 기각화 도구를 설치
- 한글 깨짐 방지: 기본 상태에서는 그래프에 한글을 쓰면 네모 모양으로 깨짐. 그래서 폰트를 설정해 주는 코드가 필수
2. 그래프의 종류와 용도
- 선 그래프(plt.plot()): 시간에 따라 데이터가 어떻게 변하는지 (예:월별 매출 추이) 볼 때 가장 좋음
- 막대 그래프(plt.bar()): 그룹 간의 크기를 비교할 때 (예: 광고 채널별 클릭수 비교) 사용
- 산점도 (plt.scatter()): 두 변수 사이의 관계나 분포 (예: 광고 비용과 매출액의 상관관계)를 점으로 찍어 볼 때 사용
- 히스토그램 (plt.hist()): 데이터가 어떤 구간에 얼마나 모여있는지 빈도 (예: 고객들의 나이대 분포)를 분석할 때 유용

```python
# [Matplotlib을 활용한 데이터 시각화 기초]

# 1. 라이브러리 설치 및 임포트
!pip install matplotlib # 설치가 안 되어 있다면 주석을 해제
import matplotlib.pyplot as plt
import numpy as np

# [필수 수칙] 한글 깨짐 방지 설정 (원도우 'Malgun Gothic', 맥 'AppleGothic')
plt.rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지

# ----- 샘플 데이터 준비 -----
months = ['1월', '2월', '3월', '4월', '5월']
sales = [1000, 1500, 1300, 2000, 2500]
channels = ['구글', '페이스북', '인스타', '유튜브']
clicks = [450, 320, 580, 700]
age_data = [22, 25, 25, 30, 32, 35, 38, 42, 45, 50, 55] # 고객 나이층 분포
```

```python
# [개념 1: 선 그래프 (Line Plot) - 매출 추이 확인]
plt.figure(figsize=(6, 4))
plt.plot(months, sales, marker='o', color='blue', linestyle='--', linewidth=2)
plt.title("월별 매출 성장 추이")
plt.xlabel("월")
plt.ylabel("매출(만원")
plt.grid(True)
plt.show()
```

```python
# [개념 2: 막대 그래프 (Bar Chart) - 채널별 클릭수 비교]
plt.figure(figsize=(6, 4))
plt.bar(channels, clicks, color='orange', alpha=0.8)
plt.title("광고 채널별 클릭수 비교")
plt.xlabel("마케팅 채널")
plt.ylabel("클릭수 (회)")
plt.show()
```

```python
# [개념 3: 산점도 (Scatter Plot) - 비용 대비 클릭수 관계]

cost = [15, 12, 18, 25, 5] # 광고 비용
actual_clicks = [450, 320, 580, 700, 150] # 실제 클릭수

plt.figure(figsize=(6, 4))
plt.scatter(cost, actual_clicks, color='red', s=100, edgecolors='black')
plt.title("광고 비용과 클릭수의 상관관계")
plt.xlabel("광고 비용 (만원)")
plt.ylabel("클릭수 (회)")
plt.grid(True)
plt.show()
```

```python
# [개념 4: 히스토그램(Histogram) - 고객 나이대 분포]

plt.figure(figsize=(6, 4))
plt.hist(age_data, bins=5, color='green', edgecolor='black', alpha=0.7)
plt.title("주요 구매 고객 나이대 분포")
plt.xlabel("나이")
plt.ylabel("고객 수 (명)")
plt.show()
```

#### 중급 핵심 문법: 도화지 쪼개기와 스타일링
1. plt.subplots() - 화면 분할하기 : 하나의 그림 창(Figure)을 바둑판처럼 나누어 여러 개의 그래프(Axes)를 동시에 그림
- 예 : plt.subplots(1, 2) -> 가로로 1중, 세로로 2칸짜리 방을 만듬
2. 스타일링 및 상세 옵션
- color & alpha: 그래프의 색상과 투명도(0~1)를 조절
- legend(): 범례(이름표)를 표시하여 각 선이나 막대가 무엇을 뜻하는지 알려줌
- tight_layout(): 그래프 간격이 좁아서 글자가 서로 겹치는 현상을 자동으로 방지해줌

```python
# [Matplotlib 중급 - 화면 분할 및 그래프 스타일링]

import matplotlib.pyplot as plt

# 처음 코드 시작시 라이브러리 설치 및 폰트 깨짐 방지 설정 (생략)

# 데이터 준비
months = ['1월', '2월', '3월', '4월', '5월']
google_clicks = [450, 520, 480, 600, 750]
fb_clicks = [320, 380, 410, 390, 420]

# 1. 1행 2열짜리 분할 화면 만들기 (가로 크기 12, 세로 크기 5)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# [첫 번째 방 - axes[0]] 구글 vs 페이스북 트렌드 비교 (선 그래프)
axes[0].plot(months, google_clicks, marker='o', color='blue', label='구글')
axes[0].plot(months, fb_clicks, marker='s', color='red', linestyle='--', label='페이스북')

axes[0].set_title("주요 채널별 클릭수 변화 추이")
axes[0].set_xlabel("월별")
axes[0].set_ylabel("클릭수 (회)")
axes[0].grid(True, linestyle=":", alpha=0.6)
axes[0].legend() # 범례(이름표) 표시

# [두 번째 방 - axes[1]] 채널별 총 비용 비교 (누적 막대 그래프)

channels = ['구글', '페이스북']
total_cost = [1200000, 850000]

# 막대 색상을 다르게 주고 투명도(alpha) 설정
axes[1].bar(channels, total_cost, color=['royalblue', 'crimson'], alpha=0.7, edgecolor='black')

axes[1].set_title("채널별 누적 광고 집행 비용")
axes[1].set_xlabel("마케팅 채널")
axes[1].set_ylabel("비용 (원)")

# 2. 그래프 간격 자동 조정 (글자 겹침 방지)
plt.tight_layout()
```

#### 고급 핵심 문법: 이중 축 활용과 이미지 저장
1. ax.twinx() 이중 y출 만들기 : 두 개의 그래프가 x축(예: 월별)은 같이 쓰지만, y축의 단위(예: 한쪽은 원, 한쪽은 퍼센트)가 완전히 다를 때 y축을 양옆(왼, 오른)으로 나누어 사용하는 기술
2. plt.savefig() 그래프 저장하기 : 주피터 노트북 화면에만 띄우는 것이 아니라 엑셀이나 파워포인트(ppt)보고서에 바로 첨부할 수 있도록 png 나 jpg 이미지 파일로 내 컴퓨터에 저장

```python
# 고급 연습 코드 (마케팅 혼합형 대시보드 리포트)
# [Matplotlib 고급 - 이중 축 구현 및 이미지 파일 저장]

import matplotlib.pyplot as plt

# 처음 코드 시작시 라이브러리 설치 및 폰트 깨짐 방지 설정 (생략)

# 실무 마케팅 데이터 준비
months = ['1월', '2월', '3월', '4월', '5월']
marketing_cost = [150, 200, 180, 300, 400] # 광고 비용 (단위: 만원)
ctr_revenue = [2.5, 3.8, 3.2, 4.5, 5.2] # 클릭률 CTR (단위: %)

# 1. 도화지와 기본 축(ax1) 만들기
fig, ax1 = plt.subplots(figsize=(8, 5))

# 2. 첫 번째 y축: 광고 비용 (막대 그래프 - 왼쪽 y축 기준)
color_cost = 'skyblue'
ax1.bar(months, marketing_cost, color=color_cost, alpha=0.7, label='광고 비용')
ax1.set_xlabel('집행 월')
ax1.set_ylabel('광고 비용 (만원)', color='steelblue')
ax1.tick_params(axis='y', labelcolor='steelblue') # 축 글자 색상 맞추기

# 3. 핵심 고급 스킬: x축을 공유하는 두 번째 y축(ax2) 만들기
ax2 = ax1.twinx()

# 4. 두 번째 y축: 클릭률 (선 그래프 - 오른쪽 y축 기준)
color_ctr = 'crimson'
ax2.plot(months, ctr_revenue, color=color_ctr, marker='o', linewidth=2, label='클릭률(CTR)')
ax2.set_ylabel('클릭률 (%)', color=color_ctr)
ax2.tick_params(axis='y', labelcolor=color_ctr)

# 5. 그래프 타이틀 및 기타 설정
plt.title("월별 광고 비용 대비 클릭률(CTR) 트렌드 분석")

# 6. 이미지 파일로 내보내기 (dpi는 화질을 뜻한다. 300이 출력용 표준!)
# bbox_inches='tight' 옵션을 주면 여백 때문에 글자가 잘리는 것을 막아줌
plt.savefig('marketing_report_day20.png', dpi=300, bbox_inches='tight')

plt.show()
print("✅ 그래프 시각화 완료 및 'marketing_report_day20.png' 파일로 저장 성공! ")
```
